## **Aproximar $f'(x_0)$** (Teoría y explicaciones) - Viernes 19 de Septiembre ✏️

Hay algunas funciones que no vamos a poder calcular su derivada de forma explicita.

En matematica pues es muy intuitivo, tenemos una $f(x)$ y queremos calcular la pendiente en un punto $x_0$, entonces explicitamente la pendiente en $x_0$ sería $f'(x_0)$

Pero realmente es rara la vez que podamos derivar explicitamente, pero lo que si podemos hacer es **aproximar $f'(x_0)$**

Primero recuerda la definición de derivada
\begin{align}
f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}
\end{align}

Entonces, una forma de aproximar $f'(x_0)$ es simplemente tomar un valor muy pequeño de $h$, esto es valido ya que la derivada es cuando $h \to 0$, entonces si agarramos un valor para $h$ muy pequeño va a ser muy buen candidato para aproximar $f'(x_0)$

\begin{align}
f'(x) \approx \frac{f(x+h)-f(x)}{h} \quad \text{para algun } h \text{ muy cercano a } 0
\end{align}

A continuación, se va a mostrar el proceso algebraico que el profe Espinoza desarrollo en la clase, en donde utiliza series de Taylor para ayudarnos a llegar a las formulas que nos van a ayudar a aproximar derivadas

### **Aproximación hacia adelante** $O(h^)$
Expandir una funcion $f(w)$ en su serie de Taylor-

\begin{align}
f(w) = &\sum_{n=0}^∞ \frac{f^{(n)} (x_0) \cdot (\color{cyan}w-x_0)^n }{n!}
\end{align}

Así, sea $w = x+h$ ⇒ $f(x+h)$ en su serie de Taylor sería:

\begin{align}
f(\underbrace{\color{cyan}{x+h}}_{\color{cyan}w} )= &\sum_{n=0}^∞ \frac{f^{(n)} (x) \cdot (\color{cyan}{x + h} - x)^n }{n!}
\end{align}

\begin{align}
f(x+h) = &\sum_{n=0}^∞ \frac{f^{(n)} (x) \cdot (h)^n }{n!}
\end{align}

Alrededor de $x_0 = x$

\begin{align}
f(x+y) = f(x) + h \cdot f'(x) + \frac{h^2}{2!} \cdot f''(x) + \frac{h^3}{3!} \cdot f^{(3)}(x) \text{ } + \text{ } ...
\end{align}

Así, $\frac{f(x+h) - f(x)}{h}$ de forma expandida queda de la forma

\begin{align}
\frac{f(x+h) - f(x)}{h} = f'(x) + \frac{h}{2!} \cdot f''(x) + \frac{h^3}{3!} \cdot f^{(3)}(x) \text{ } + \text{ } ...
\end{align}

#### Explicación algebraica

Talvez te estas preguntando ¿por que $\frac{f(x+h) - f(x)}{h}$ queda de esa forma?

Basicamente agarramos la forma expandida de $f(x+h)$ que se mostro anteriormente, luego de esa forma expandida le restamos la expresión $f(x)$ en ambos lados, entonces nos queda:

\begin{align}
f(x+y) = \color{cyan}{f(x)} + h \cdot f'(x) + \frac{h^2}{2!} \cdot f''(x) + \frac{h^3}{3!} \cdot f^{(3)}(x) \text{ } + \text{ } ...
\end{align}

\begin{align*}
f(x+h) \color{red}{- f(x)} &= \color{cyan}{f(x)} \color{red}{-f(x)} + h \cdot f'(x) + \frac{h^2}{2!} f''(x) + \dots
\end{align*}

\begin{align}
f(x+h) - f(x) = h \cdot f'(x) + \frac{h^2}{2!} \cdot f''(x) + \frac{h^3}{3!} \cdot f^{(3)}(x) \text{ } + \text{ } ...
\end{align}

Luego dividimos ambos lados entre $h$, y así es como nos queda la ecuación

\begin{align*}
\frac{f(x+h) - f(x)}{\color{red}{h}} &= \frac{\color{cyan}{h} \cdot f'(x)}{\color{red}{h}} + \frac{\color{cyan}{h^2} \cdot f''(x)}{2! \cdot \color{red}{h}} + \frac{\color{cyan}{h^3} \cdot f'''(x)}{3! \cdot \color{red}{h}} + \dots \
\end{align*}

Y así nos queda la ecuación de la forma
\begin{align}
\frac{f(x+h) - f(x)}{h} = f'(x) + \frac{h}{2!} \cdot f''(x) + \frac{h^3}{3!} \cdot f^{(3)}(x) \text{ } + \text{ } ...
\end{align}

#### ¿Que significa la ecuación anterior?

La ecuación anterior es clave para poder aproximar derivadas en donde cada factor es importante:

$$
\underbrace{\frac{f(x+h) - f(x)}{h}}_{\text{Derivada aproximada}}
=
\underbrace{f'(x)}_{\text{El Valor Exacto}}
+ \underbrace{\frac{h}{2!} f''(x) + \frac{h^2}{3!} f^{(3)}(x) + \cdots}_{\text{El Error } O(h)}
$$

- **La Fórmula de Aproximación:** El lado izquierdo, conocido como la fórmula de diferencia hacia adelante, es la herramienta que usaremos para calcular nuestra aproximación numérica. Se le llama "hacia adelante" porque evalúa la función en el punto actual (x) y en un punto un "pequeño paso" hacia adelante (x+h).

- **El Valor Exacto:** Es el valor real de la derivada que queremos encontrar: $\text{valor exacto} := f'(x_0)$
- **El Error:** Todos los términos restantes representan la diferencia entre nuestra aproximación y el valor exacto. Este se conoce como error de truncamiento o tambien denotado como $O(h)$, porque "truncamos" o cortamos la serie infinita.

Como el primer término del error, $\frac{h}{2!} f''(x)$ depende de $h$, decimos que el error es de **orden $h$**, o simplemente $O(h)$. Esto significa que si reducimos h a la mitad, el error también se reducirá aproximadamente a la mitad.

### **Aproximación central** $O(h^2)$

Ya hemos calculado el "paso hacia adelante", pero hay otro método aun mas eficiente que ahora tambien da un "paso hacia atras"

Esta es la método de diferencia central. Geométricamente, calcula la pendiente de la línea secante que pasa por los puntos $(x−h,f(x−h))$ y $(x+h,f(x+h))$. Esta línea suele ser una mejor aproximación que el método de "hacia adelante".

\begin{align}
f(\underbrace{\color{cyan}{x-h}}_{\color{cyan}w} )= &\sum_{n=0}^∞ \frac{f^{(n)} (x) \cdot (\color{cyan}{x - h} - x)^n }{n!}
\end{align}

\begin{align}
f(x+h) = &\sum_{n=0}^∞ \frac{f^{(n)} (x) \cdot (-h)^n }{n!}
\end{align}

\begin{align}
f(x+h) = &\sum_{n=0}^∞ \frac{ (-1)^n \cdot f^{(n)} (x) \cdot (h)^n }{n!}
\end{align}

Alrededor de $x_0 = x$

\begin{align}
f(x-h) = f(x) - hf'(x) + \frac{h^2}{2!}f''(x) - \frac{h^3}{3!}f^{3}(x) + \frac{h^4}{4!}f^{4}(x) \text{ } - \text{ } ...
\end{align}

Así,
\begin{align}
f(x+y) = f(x) + h \cdot f'(x) + \frac{h^2}{2!} \cdot f''(x) + \frac{h^3}{3!} \cdot f^{(3)}(x) \text{ } + \text{ } ...
\end{align}

\begin{align}
f(x-h) = f(x) - hf'(x) + \frac{h^2}{2!}f''(x) - \frac{h^3}{3!}f^{3}(x) + \frac{h^4}{4!}f^{4}(x) \text{ } - \text{ } ...
\end{align}

Entonces,
\begin{align}
f(x+h) - f(x-h) = \color{red}{f(x) - f(x)} + hf'(x) - [-hf'(x)] + \color{red}{\frac{h^2}{2!}f''(x) - \frac{h^2}{2!}f''(x)} + \frac{h^3}{3!} f^{(3)}(x) - [-\frac{h^3}{3!}f^{(3)}(x)] + \color{red}{ \frac{h^4}{4!}f^{(4)}(x) - \frac{h^4}{4!}f^{(4)}(x) } \text{ } + \text{ } ...
\end{align}

Al calcular $f(x+h) - f(x-h)$ hay un monton de terminos que se cancelan y calcularlo a mano puede ser un poco tardado, asi que resumidamente quedaría como:

\begin{align}
f(x+h) - f(x-h) = 2hf'(x) + 2\frac{h^3}{3!} f^{(3)} + 2\frac{h^5}{5!} f^{(5)} + 2\frac{h^7}{7!} f^{(7)} \text { } + \text { } ...
\end{align}

Dividiendo de ambos laods por $2h$, quedaría como

\begin{align}
\frac{f(x+h) - f(x-h)}{2h} = f'(x) + O(h^2)
\end{align}

Asi es como se concluye este apartado

En resumen, obtuvimos dos formulas para aproximar la derivada en un punto $f'(x_0)$, las formulas en cuestion son:

- Aproximación hacia adelante:
\begin{align}
\frac{f(x+h) - f(x)}{h} = f'(x) + O(h)
\end{align}

- Aproximación central:
\begin{align}
\frac{f(x+h) - f(x-h)}{2h} = f'(x) + O(h^2)
\end{align}

NOTA: recuerda que $O(h^n)$ representa solo el error de aproximación.


## **Aproximar derivadas en un punto $x_0$** ($O(h), O(h^2), O(h^4)$) - Lunes 22 de Septiembre 📑

### **Instrucciones del ejercicio ✏️**

Sea $f(x) = \sin (x) + x \cdot \cos (x)$

1.  Calcular $f'(x)$ simbolicamenye y evaluar en $x = \frac{π}{2}$

    " Valor exacto": $f'(\frac{π}{2})$ (radianes)

\begin{align}
D_h = \frac{f(x+h) - f(x-h))}{2h}
\end{align}

2.  Aproximar $f'(\frac{π}{2})$ hasta una precisión menor que `10^(-9)` con las siguientes aproximaciones:

    (i)
    \begin{align}
    f'(x_0) = \frac{f(x_0 + h) - f(x_0)}{h} + O(h)
    \end{align}

    (ii)
    \begin{align}
    f'(x_0) = \frac{f(x_0 + h) - f(x_0 - h)}{2h} + O(h^2)
    \end{align}

    (iii)
    \begin{align}
    f'(x_0) = \frac{4D_{h/2} - D_h}{3} + O(h^4)
    \end{align}

    con:
  
  \begin{align}
  \quad \quad h = \{1,0.5,0.25,0.0125\}
  \end{align}

  \begin{align}
    O(h^n): \quad |\text{Error}| ≤ \text{constante} \cdot h^n, \quad h << 1
  \end{align}
    

### **Código: Aproximar derivadas con 3 métodos ($O(h), O(h^2), O(h^4)$)** `👨‍💻`

````python
import sympy as sp
import numpy as np
import pandas as pd

# --------------------------------------------------------------------------
# PASO 1: Calcular la derivada simbólica y el valor exacto
# --------------------------------------------------------------------------

x = sp.Symbol('x')
f_sym = sp.sin(x) + x * sp.cos(x) # La función simbólica
f_prime_sym = sp.diff(f_sym, x)   # Derivada simbólica
x0 = sp.pi / 2                    # El punto en el que queremos evaluar la derivada
exact_value = f_prime_sym.subs(x, x0).evalf() # El valor exacto de f'(x_0)

print("--- Análisis Simbólico ---")
print(f"La función f(x) es:")
display(f_sym)
print("\nLa derivada f'(x) es:")
display(f_prime_sym)
print(f"\nEl valor exacto de f'({x0}) es: {exact_value}")
print("="*30)

# --------------------------------------------------------------------------
# PASO 2: Realizar las aproximaciones numéricas (La tabla de aproximaciones)
# --------------------------------------------------------------------------

f_num = sp.lambdify(x, f_sym, 'numpy') # convertimos f(x) en una funcion de python normal de tipo lambda
x0_num = float(x0) # Convertimos pi/2 a un numero flotante 1.57079...
h_values = [1, 0.5, 0.25, 0.0125] # Lista para los valores que toma h
results = [] # Se usara esta lista para guardar los resultados de CADA FILA

# Funcion para la aproximación (i)
def i_derivada(f, x0, h):
    return (f(x0 + h) - f(x0)) / h

# Funcion para la aproximación (ii)
def ii_derivada(f, x0, h):
    return (f(x0 + h) - f(x0 - h)) / (2 * h)

# Recorremos cada h, h = 1, h = 0.5, h = 0.25, ...
for h in h_values:
    # (i) Aproximación hacia adelante
    i_valor = i_derivada(f_num, x0_num, h)           # Valor aproximado
    i_error = abs(i_valor - exact_value)   # Margen de error

    # (ii) Aproximación central
    ii_valor = ii_derivada(f_num, x0_num, h)
    ii_error = abs(ii_valor - exact_value)

    # (iii) Aproximación de orden superior
    D_h2 = ii_derivada(f_num, x0_num, h / 2)
    approx_iii = (4 * D_h2 - ii_valor) / 3
    error_iii = abs(approx_iii - exact_value)

    # La variable approx_iii esta definida por ( 4*D_{h/2} - D_h )/3
    # Pero como ya vimos en las notas, D_h es lo mismo que lo que se obtiene en ii_valor

    # Cambiamos los nombres de las claves para que coincidan con los que pediste
    results.append({
        'h': h,
        '(i)': i_valor,
        '|E_1|': i_error,
        '(ii)': ii_valor,
        '|E_2|': ii_error,
        '(iii)': approx_iii,
        '|E_3|': error_iii
    })

df = pd.DataFrame(results)

# Establecemos la columna 'h' como el índice de la tabla
df = df.set_index('h')

pd.options.display.float_format = '{:.10e}'.format

print("\n--- Tabla de Aproximaciones y Errores ---")
display(df)
````

# **Integración númerica** 📌

## **Introducción a los métodos de integración** - Martes 23 de septiembre del 2025 ✏️

La integración numérica se utiliza para aproximar el valor de una integral definida $\int_a^bf(x)dx$ cuando no es posible o práctico calcularla de forma analítica. La idea fundamental es dividir el área bajo la curva en formas geométricas simples (rectángulos, trapecios, etc.) y sumar sus áreas.

Vamos a ver 3 métodos en total:
- Suma de reimann con rectangulos izuqierdo o derechos
- punto medio (variante de suma de reimann)
- trapecios
- Regla de Simpson

La definición formal de la integral es el límite de una suma:

$$
\int_a^b f(x)\,\mathrm{d}x
= \lim_{n \to ∞} \sum_{i=1}^n f(x_i) \,\Delta x, \quad \text{donde } \Delta x = \frac{b-a}{n} \\
$$

En la integración numérica, no tomamos el límite, sino que usamos un número finito $N$ de formas, lo que introduce un error de aproximación.

### **1. Sumas de Riemann (Aproximación con Rectángulos)** 📏

Este es el método más fundamental. Aproximamos el área con una serie de rectángulos de ancho $Δx$.

La altura de cada rectángulo se puede determinar por el valor de la función en el extremo derecho ($x_i$) o izquierdo ($x_{i−1}$) de su base.

$$
\int_a^b f(x)\,\mathrm{d}x ≈ \sum_{i=1}^n f(x_i) \,\Delta x + O(\Delta x)
$$

$$
x_i \text{ esta a la derecha de cada rectangulo o bien a la izquierda}
$$

- **Error:** La precisión de este método es limitada. El error es de orden $Δx$, lo que se denota como $O(Δx)$. Esto significa que si duplicamos el número de rectángulos, el error se reduce aproximadamente a la mitad.

### **2. Regla del punto medio** 📏

Esta es una mejora de la Suma de Riemann. En lugar de usar los extremos, la altura de cada rectángulo se determina por el valor de la función en el punto medio de su base, $\bar{x}_i = \frac{x_{i-1} + x_i}{2}$

$$
\int_a^b f(x)\,\mathrm{d}x ≈ \sum_{i=1}^n f(\bar{x}_i) \,\Delta x + O((\Delta x)^2) \\
\text{Con } x_i \text{ el punto de medio de la base de cada rectangulo}
$$
- Error: Este método es considerablemente más preciso. El error es de orden $(Δx)^2$, denotado como $O((Δx)^2)$. Si duplicamos el número de rectángulos, el error se reduce a una cuarta parte.

### **3. Regla del trapecio** 📏

En lugar de rectángulos, este método aproxima el área usando trapecios, lo que se ajusta mejor a la forma de la curva

Para construr

El área de un solo trapecio (desde $x_i$ hasta $x_{i+1}$ es $\frac{f(x_i) + f(x_{i+1})}{2} \cdot Δx$

$$
\int_a^b f(x)\,\mathrm{d}x ≈ \frac{f(x_0) + f(x_1)}{2} \cdot \Delta x\\
+ \frac{f(x_1) + f(x_2)}{2} \cdot \Delta x \\
+ \frac{f(x_2) + f(x_3)}{2} \cdot \Delta x \\
\text{Lo anterior es una susecion, escrito de forma de sumatoria sería:}\\
\int_a^b f(x)\,\mathrm{d}x ≈  \sum_{i=1}^{n-1} \frac{f(x_i) + f(x_{i+1})}{2} \cdot \Delta x \\
\text{Entonces, podemos factorizar } \frac{\Delta x}{2} \text{, así nos queda y concluimos:}\\
∴\int_a^b f(x)\,\mathrm{d}x ≈ \frac{\Delta x}{2} [f(x_0) + 2f(x_1) + 2f(x_2 + ... + 2f(x_{n-1}) + f(n)] + O((Δx)^2)
$$

- **Error:** La precisión es similar a la del punto medio. El error es de orden $(Δx)^2$, es decir, $O((Δx)^2)$.

### **4. Regla de Simpson (Aproximación con Parábolas)** 📏

Este método lleva la aproximación un paso más allá, utilizando parábolas para conectar conjuntos de tres puntos, lo que permite seguir curvas de forma mucho más precisa.

Para construir la formula, primero calculamos el area bajo una sola parabola $p(x)$ que pasa por tres puntos $(x_0, f(x_0)), (x_1, f(x_1)), (x_2, f(x_2))$

A continuación se va a mostrar el desarrollo algebraico que se desarrollo en clase, mas tarde vamos a darle una intrepretación final.

Sea $p(x)$ la parabola con forma $Ax^2 + Bx + C$

$$
\int_a^b f(x)\,\mathrm{d}x ≈ \int_a^b p(x)\,\mathrm{d}x ⇒ \\
\text{donde } p(x) \text{ podemos expresarlo como una parabola } p(x) = Ax^2 + Bx + C \\
\int_a^b (Ax^2 + Bx + C)\,\mathrm{d}x \\
= \frac{Ax^3}{3} + \frac{Bx^2}{2} + Cx |_a^b
$$

Integración & optimización numérica (3).png

Vamos a agarrar $p(x) = Ax^2 + Bx + C$
$$
p(0) = A(0)^2 + B(0) + C = C = f(\frac{a+b}{2}) \\
p(h) = Ah^2 + Bh + c = f(b) \\
p(-h) = A(-h)^2 + B(-h) + C = Ah^2 - Bh + C = f(a)\\
\text{Entonces, } A,B,C = ?
$$

El proceso para encontrar $A,B,C$ es complicado algebraicamente, en resumen queda:

$$
A = \frac{ f(a) + f(b) - 2f(\frac{a+b}{2}) }{ 2h^2 } \\
B = \frac{f(b) - f(a)}{2h} \\
C = f(\frac{a+b}{2})\\
\text{Ahora integramps:}
\int_{-h}^{h} (Ax^2 + Bx + C) dx =\\
\left. \frac{Ax^3}{3} \right|_{-h}^h = \frac{2Ah^3}{3} +\\
\left. \frac{Bx^2}{2} \right|_{-h}^h = 0 \\
\left. \frac{Cx}{1} \right|_{-h}^h = 2Ch\\
= \frac{2Ah^3}{3} + 2Ch \\
\text{⇒ Sustituyendo } A,B, \text{ y } C \\
= \frac{ \color{red}{2h}^3}{3} \cdot \frac{ f(a) + f(b) - 2f(\frac{a+b}{2}) }{ \color{red}{2h}^2 }   + 2h \cdot f(\frac{a+b}{2}) \\
\text{Se cancela } 2h^2 \text{ del nominador con } 2h^2 \text{ del denominador} \\
= \frac{ \color{red}{2h}^3}{3} \cdot \frac{ f(a) + f(b) - 2f(\frac{a+b}{2}) }{ \color{red}{2h^2} }   + 2h \cdot f(\frac{a+b}{2}) \\
= \frac{h}{3} \cdot [ f(a) + f(b) - 2f(\frac{a+b}{2}) ] + \color{cyan}{2h} \cdot f(\frac{a+b}{2}) \\
\text{expresamos } 2h \text{ como } \frac{6}{3}h \\
= \frac{h}{3} \cdot [ f(a) + f(b) - 2f(\frac{a+b}{2}) ] + \frac{6}{3}h \cdot f(\frac{a+b}{2}) \\
\text{Simplificando y tomando } \frac{h}{3} \text{ como factor común} \\
= \frac{h}{3} \cdot [ f(a) + f(b) + 4f(\frac{a+b}{2}) ] \\
\text{Tomando } h = \frac{b-a}{2} \\
= \frac{b - a}{6} \cdot [ f(a) + f(b) + 4f(\frac{a+b}{2}) ] \\
$$

**Finalización de las notas de clase**

Para poder integrar en un intervalo $[a,b]$, necesitamos dividir este intervalo en zonas mas pequeñas, vamos a denotarlo como $[x_0, x_1, x_2, ..., x_n]$ el intervalo dividido en varios pedazos mas pequeños

Entonces, el area bajo la curva entre $[x_0, x_2]$ sería:

$$
\int_{x_0}^{x_2} f(x) \ dx ≈ \int_{x_0}^{x_2} p(x) \ dx = \frac{\Delta x}{3} [f(x_0) + 4f(x_1) + f(x_2)] \\
$$

Ahora, para encontrar la integral completa en un intervalo $[a,b]$, aplicamos la formula anterior repetidamente a lo largo de la curva. Sumamos el área del primer segmento parabólico (de $x_0$ a $x_2$), luego el segundo (de $x_2$ a $x_4$), y asi sucesivamente.

$$
\int_a^b f(x) \, dx = \underbrace{ \frac{\Delta x}{3} \left[ f(x_0) + 4f(x_1) + f(x_2) \right]}_{Área 1} + \underbrace{\frac{\Delta x}{3} \left[ f(x_2) + 4f(x_3) + f(x_4) \right]}_{Área 2} + \cdots
$$

Así, la formula final sería:
$$
\int_a^b f(x) \, dx ≈ \frac{\Delta x}{3} \left[ f(x_0) + 4f(x_1) + 2f(x_2) + 4f(x_3) + 2f(x_4) + \cdots + 4f(x_{n-1}) + f(x_n) \right] + O((Δx)^4)
$$

**Importante:** Para que esta regla funcione, el número de subintervalos $n$ debe de ser par.
- **Error:** La regla de Simpson es extremadamente precisa. El error es de orden $(Δx)^4$, es decir, $O((Δx)^4)$. Si duplicamos el número de intervalos, el error se reduce por un factor de 16, lo que lo hace demasiado preciso

### **Sumas de Reiman (Izquierda, derecha y punto medio)** 📏

El método de las sumas de Reiman tienen 3 variaciones para poder calcular los rectangulos: por la izquierda, derecha y por en medio.

$$
\text{izquierda: } \int_a^b f(x) \ dx ≈ h \cdot \left[ f(a) + f(a+h) + f(a + 2h) + \cdots + f(a+5h)  \right] \\
\text{medio: } \int_a^b f(x) \ dx ≈ h \cdot \left[ f(a + \frac{h}{2}) + f(a + \frac{h}{2} + h) + f(a + \frac{h}{2} + 2h) + \cdots + f(a + \frac{h}{2} + 5h) \right] \\
\text{derecha: } \int_a^b f(x) \ dx ≈ h \cdot \left[ f(a + h) + f(a+2h) + f(a + 3h) + \cdots + f(a+6h)  \right] \\
$$

Algoritmicamente, serría algo como:
```
# 1. Inicializar la suma a cero ANTES de empezar el bucle y definimos h
Suma_alturas = 0
N = 6
h = (b - a) / N

# 2. Iniciar el bucle para recorrer cada rectángulo
for i: 0 → N-1    # En python sería for i in range(0, N)
  # Calcular la altura del rectángulo actual según el método
  if izquierda:
    altura_actual = f(a + i*h)
  if medio:
    altura_actual = f(a + i*h + h/2)
  if derecha:
    altura_actual = f(a + (i+1)*h)

  # 3. AÑADIR (acumular) la altura actual a la suma total
  Suma_alturas += altura_actual # Esto es lo mismo que: Suma_alturas = Suma_alturas + altura_actual

# 4. Calcular el área total DESPUÉS de que el bucle haya terminado
Area_aproximada = Suma_alturas * h
```

### **Regla del trapecio** 📏

La lógica es similar a la de Riemann, pero en lugar de sumar áreas de rectángulos, sumaremos las áreas de trapecios.

$$
\int_a^b f(x) \ dx ≈ \frac{h}{2} \cdot [f(a) + 2f(a+h) + f(a+2h) + \cdots + f(b)]
$$

Algoritmicamente sería algo como:
```
# 1. Inicializar la suma a cero ANTES de empezar el bucle y definimos h
Suma_alturas = 0
N = 6
h = (b - a) / N

# 2. Si te fijas en la formula, el primer y el ultimo término en la suma no repiten un patron, así que es mejor sumarlos de una vez
Suma_alturas = f(a) + f(b)

# 3. Iniciar el bucle para la sumatoria
for i: 1 → N-1    # En python sería for i in range(1, N)
  Suma_alturas += 2 * f(a + h*i)


# 4. Calcular el área total DESPUÉS de que el bucle haya terminado
Area_aproximada = Suma_alturas * (h/2)
```

### **Regla de Simpson** 📏

La lógica de esta formula es muy parecida a la de la regla del trapecio, pero los coeficientes de la suma ahora son diferentes, siguiendo el patrom `1, 4, 2, 4, 2, 4, 2, .., 4, 1`.

Recuerda que para este método, es necesario que $N$ sea un número par:

$$
\int_a^b \ dx ≈ \frac{h}{3} \cdot [f(a) + 4f(a+h) + 2f(a+2h) + 4f(a+3h) + \cdots + f(b)]
$$

El algoritmo sería algo como:
```
# 1. Verificar que N es par
if N no es par:
  Mostrar error y detener
  
# 2. Definir h e inicializar la suma con los extremos
h = (b - a) / N
Suma_terminos = f(a) + f(b)

# 3. Recorrer los puntos interiores (de 1 a N-1)
for i: 1 → N-1
  
  # 4. Aplicar el coeficiente 4 para índices IMPARES y 2 para PARES
  if i es IMPAR:
    Suma_terminos += 4 * f(a + i * h)
  else: # si i es PAR
    Suma_terminos += 2 * f(a + i * h)

# 5. Calcular el área final DESPUÉS del bucle
Area_aproximada = Suma_terminos * (h / 3)
```

````python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import pandas as pd

t = sp.Symbol('t')

# La función de sumas de Reimann
def IntReimann(f, a, b, N, metodo='izquierda') :

    h = (b - a) / N
    suma_alturas = 0.0

    for i in range(N):
        if metodo == 'izquierda':
            punto_x = a + i * h
        elif metodo == 'derecha':
            punto_x = a + (i + 1) * h
        elif metodo == 'medio':
            punto_x = a + i * h + h / 2
        else:
            raise ValueError("Método no reconocido. Use 'izquierda', 'derecha' o 'medio'.")

        suma_alturas += f(punto_x)

    AreaAprox = suma_alturas * h
    return AreaAprox

# Integral con trapecios
def IntTrapecios(f, a, b, N) :

    h = (b - a) / N

    # Inicializamos la suma
    suma_alturas = f(a) + f(b)

    for i in range(1, N):
        punto_x = a + i * h
        suma_alturas += 2 * f(punto_x)

    AreaAprox = suma_alturas * (h/2)
    return AreaAprox

# Integral con simpson, NOTA IMPORTANTE: N debe de ser par
def IntSimpson(f, a, b, N) :
  if N % 2 != 0 :
    raise ValueError("El número de intervalos N debe de ser par.")

  h = (b - a) / N

  # Inicializar la suma con los extremos f(x_0) y f(x_n) que tambien es f(a) y f(b)
  suma = f(a) + f(b)

  # Sumar los puntos que hay entre (a,b)
  for i in range(1, N):
    punto_x = a + i * h
    if i % 2 == 0: # Si el indice "i" es par:
      suma += 2 * f(punto_x)
    else: # Si el indice "i" es impar
      suma += 4 * f(punto_x)

  # Multiplicar la suma total por (h/3)
  AreaAprox = (h / 3) * suma
  return AreaAprox

# Definir la función f_num(val) usando las funciones de numpy (np)
def f_num(val):
    return 50 + 30 * np.sin((np.pi * val) / 12) + 15 * np.cos((np.pi * val) / 6)

# 2. Función SIMBÓLICA (para la integral exacta)
f_sym = 50 + 30 * sp.sin((sp.pi * t) / 12) + 15 * sp.cos((sp.pi * t) / 6)


# ===============================
# MAIN
# ===============================

a = 0
b = 24
# --- Parte (a) ---
N = 6

AreaReiman = IntReimann(f_num, a, b, N, metodo='izquierda')
AreaTrapecios = IntTrapecios(f_num, a, b, N)
AreaSimpson = IntSimpson(f_num, a, b, N)

print("--- Parte (a): Aproximaciones con N = 6 ---")
print(f"Suma de Riemann (izquierda): {AreaReiman:.4f} GB")
print(f"Regla del Trapecio:          {AreaTrapecios:.4f} GB")
print(f"Regla de Simpson:            {AreaSimpson:.4f} GB")

# --- Parte (b) ---
# Usamos sympy.integrate con la función simbólica f_sym
valor_exacto = sp.integrate(f_sym, (t, a, b))
valor_exacto = float(valor_exacto) # Convertir a número

print("\n--- Parte (b): Valor Analítico ---")
print(f"El valor exacto de la integral es: {valor_exacto:.4f} GB")
print(f"Comparando con los otros resultados, es exactamente el mismo área.")

# --- Parte (c) ---
N_list = [2, 4, 6, 8, 10, 12]
errors_riemann = []
errors_trapecio = []
errors_simpson = []

for N_c in N_list:
    aprox_riemann = IntReimann(f_num, a, b, N_c, metodo='izquierda')
    aprox_trapecio = IntTrapecios(f_num, a, b, N_c)
    aprox_simpson = IntSimpson(f_num, a, b, N_c)

    errors_riemann.append(abs(aprox_riemann - valor_exacto))
    errors_trapecio.append(abs(aprox_trapecio - valor_exacto))
    errors_simpson.append(abs(aprox_simpson - valor_exacto))

# Mostrar tabla de errores
df_errors = pd.DataFrame({
    'N': N_list,
    'Error Riemann (Izq)': errors_riemann,
    'Error Trapecio': errors_trapecio,
    'Error Simpson': errors_simpson
})
df_errors.set_index('N', inplace=True)
pd.options.display.float_format = '{:.10e}'.format

print("\n--- Parte (c): Tabla de Errores Absolutos ---")
print(df_errors)

# Graficar los errores
plt.figure(figsize=(10, 6))
plt.plot(N_list, errors_riemann, 'o-', label='Error Riemann (Izquierda)')
plt.plot(N_list, errors_trapecio, 's-', label='Error Trapecio')
plt.plot(N_list, errors_simpson, '^-', label='Error Simpson')

plt.xlabel("Número de Intervalos (N)")
plt.ylabel("Error Absoluto (Escala Logarítmica)")
plt.title("Error de Integración vs. Número de Intervalos")
plt.legend()
plt.grid(True)
plt.yscale('log') # Escala logarítmica es mejor para ver errores pequeños
plt.xticks(N_list)
plt.show()
````

### **Codigo para encontrar el maximo volumen con Monte-Carlo & epsilons:** 👨‍💻

````python
import sympy as sp
import random

x = sp.Symbol('x')

def v(x) :
  return 400*x - 80*x**2 + 4*x**3

def vmax():
  vmax = 0

  for i in range(0, 10001): # Itera 10000 veces
    aleatorio = random.uniform(0, 10) # Calcula algun numero aleatorio entre 0 y 10

    candidato = v(aleatorio) # Entonces, ese numero aleatorio es nuestro candidato para ser el maximo local
    if candidato > vmax : #  Si nuestro candidato es mejor que nuestra anterior mejor aproximacion, entonces nuestro candidato para a ser la mejor aproximacion
      vmax = candidato
  return vmax


def metodoepsilon ():
  aleatorio = random.uniform(0,10) # Generamos un numero aleatorio entre 0,10
  epsilon =10**(-3) # Un valor muy pequeño epsilon
  algo = 0 # Una maña para poner while (True) XDDD
  x_0 = aleatorio # x_0 es un aleatorio

  while (algo == 0) : # while true xdddd
    plus =x_0 + epsilon # Nos movelos un poquito a la derecha para poder evaluar
    menos = x_0 - epsilon # Nos movemos un poquito a la izquierda para poder evaluar

    if(v(x_0)>= v(plus) and v(x_0)>= v(menos)): # Comprobamos si v(x_0) ha converguido al maximo local, y si lo es entonces se rompe el while true
      break

    if( v(x_0) < v(menos) ): # Si v(x_0 - epsilon) es maypr a v(x_0), entonces la funcion crece hacia la izquierda, por lo tanto ahora asignamos x_0 = x_0 - epsilon
      x_0 = menos
    else: # De lo contrario, significa que la funcion crece hacia la derecha, entonces x_0 = x_0 + epsilon
      x_0 = plus
  return v(x_0)


vmax = vmax()
maximoLocal = metodoepsilon()
print("vmax: ",vmax)
print("maximo local: ", maximoLocal)
````

### **Descenso por gradiente** ✏️

La clase de hoy es como lo que vimos ayer, hoy vamos a ver otro método para poder aproximar el maximo.

**Descenso por gradiente**

Este método tambien convergue hacía el maximo local así como la función `metodoEpsilon`

- Los parametros iniciales son `v(x)`, `v'(x)` y un $α$ que estará definido como `α := 10^(-3)`.

- Agarramos un aleatorio $x_0 ∈ [0,10]$

- Ahora el chiste es encontrar $x_1$, el cual esta definido como: $x_1 = x_0 + α \cdot v'(x_0)$

- En general, $x_{n+1} = x_n + α \cdot v'(x_n)$

La idea de esto es que $v'(x_n)$ nos guía ayuda a guiarnos para encontrar el maximo local.

Digamos que para nuestra primera iteración, en $x_0$ nos da una pendiente negativa, eso nos indica que estamos en el lado por donde la funcion esta decreciendo, así que ahora nos vamos a mover hacía la izquierda.

Si el caso anterior pasa, entonces el factor `α * v'(x_0)` va a ser negativo, entonces `x_1` va a estar ligeramente mas hacía la izquierda, lo cual es correcto para acercarnos al punto maximo.

Digamos que de pura casualidad $x_0$ ya esta en el máximo local, eso significa que $v'(x_0) = 0$, por lo tanto $x_1 = x_0 + 0$ ¡ya no nos movemos por que estamos en el maximo local!

como nota, el profe denoto $α$ como **la tasa de aprendizaje**, supongo que este es un detalle muy importante para mas adelante en la carrera...

#### **¿Como estimar π?** ✏️

Digamos que Tenemos un cuadrado de area $2×2$ y adentro de dicho cuadrado hay un disco $D$ con radio 1...

image.png

$$
A_{\text{circulo}} = π(1)^2 = π \\
A_{\text{cuadrado}} = 2\cdot2 = 4 \\
\text{Lanzar: } N \text{ puntos} \quad \text{(N = 10,000)}
$$

El dilema consta en lanzar $N$ puntos aleatorios entre $[-1,1]$, ahora nos preguntamos si ese punto $P$ esta adentro del circulo, si esta entonces es un caso exitoso, y si no lo esta pues es un fracaso...

¿Cual es la probabilidad de que ese punto este adentro del disco?

$$
\text{probabilidad} (P ∈ D ) = \frac{π}{4} \\
\text{"La probabilidad de que un punto este en el disco es de pi/4"}
$$

Entonces, digamos que iteramos un total de 10,000 veces y que obtuvimos 1,000 casos exitosos en donde $P$ esta el disco, entonces la ecuación sería:

$$
\frac{1000}{10000} ≈ \frac{π}{4} \\
π ≈ 0.4
$$

Así, la formula para aproximar $π$ sería:

$$
π ≈ \frac{4 ⋅ \text{#Exitos}}{N}
$$

image.png

Ahora la pregunta es, ¿Como determinar si un punto $P$ es exitoso?

Este es un problema geometrico, la logica que hay detras para determinar si P es exitoso o no es calculando la distancia que hay desde el centro de la circuferencia hasta el punto $P$, así. Sabemos que el radio del disco es de 1, así que si la distancia del centro al punto $P$ es mayor que uno, quiere decir que $P$ NO esta adentro del circulo, por lo tanto es un caso fracaso.

  exitos = exitos + 1 # Le sumamos

#### **Código: Estimar pi con Monte-Carlo** 👨‍💻

````python
# Codigo hecho por Axel e Ivana
import sympy as sp
import random

def distancia(x, y) :
  return (x**2 + y**2)**(1/2)

N = 10000
exitos = 0

for i in range(0, N) :

  punto_x = random.uniform(-1,1)
  punto_y = random.uniform(-1,1)

  if distancia(punto_x , punto_y) <= 1 :
    exitos += 1

pi_aprox = (4 * exitos)/N


print("Aproximacion: ", pi_aprox)
````

El proceso se repite durante varias **épocas**. Es fundamental barajar (shuffle) los datos al inicio de cada época.

## **Práctica de análisis numérico - Optimización: GD vs SGD para regresión lineal** - Ejecución 📌📝

### Codigo preparativo 📑👨‍💻

````python
# ===== PARTE 1 ===

import numpy as np
import matplotlib.pyplot as plt
import time

# Fijamos la semilla para asegurar la reproducibilidad de los resultados
np.random.seed(20251015)

# 1. Crear el array x
# Usar np.linspace(inicio, fin, numero_de_puntos) para asegurar 101 puntos.
x = np.linspace(0, 10, 101)

# 2. Crear el array de ruido
# Usar np.random.uniform(low, high, size)
noise = np.random.uniform(-1, 1, size=101)

# 3. Crear el array y
w1_true = 8.0
w2_true = 5.0
y = w1_true + w2_true*x + noise

N = len(x)
'''
print(f"Número de puntos generados (N): {N}")
if N != 101:
    print("WTF... el número de puntos no es 1001!")
'''
plt.figure(figsize=(10, 6))
# Descomenta las siguientes líneas después de generar x e y
plt.scatter(x, y, s=10, alpha=0.6, label='dataset')
# Graficamos la recta (sin ruido) para referencia
plt.plot(x, w1_true + w2_true*x, color='red', linewidth=2, label='Recta (y=8+5x)')
plt.title("Dataset")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()

# === PARTE 2 ===

# 2.1 El modelo (hipotesis)
def predict(x, w1, w2):
    """Calcula la predicción de la regresión lineal.
    La fórmula es: y_hat = w1 + w2 * x
    """
    return w1 + w2 * x

# 2.2. Funcion objetivo Q(w) (suma de errores cuadrados SSE)
def compute_cost(y_pred, y_true):
    """Calcula la función objetivo Q(w) (SSE)."""
    error = y_pred - y_true
    # Calcula los errores al cuadrado
    squared_errors = error ** 2
    # Suma los errores (¡No dividir por N!)
    total_cost = np.sum(squared_errors)
    return total_cost

# === PARTE 3 ===

# 3.1 Cálculo del gradiente
def compute_gradients_batch(x, y_true, y_pred):
    """Calcula los gradientes totales para w1 y w2 (Batch)."""
    # 1. Calcular el error (predicción - real)
    error = y_pred - y_true

    # 2. Gradiente para w1
    grad_w1 = np.sum(2 * error)

    # 3. Gradiente para w2
    grad_w2 = np.sum(2 * error * x)

    return grad_w1, grad_w2

# 3.2 Implementación del algoritmo GD
def batch_gradient_descent(x, y, eta, num_iterations):
    # 1. Inicialización de parámetros
    w1 = 0.0
    w2 = 0.0
    # Historia para registrar la evolución
    history = {'cost': [], 'w1': [], 'w2': []}

    print("\n--- Iniciando Batch Gradient Descent (GD) ---")
    start_time = time.time()

    # 2. Ciclo de optimización
    for i in range(num_iterations):
        # a. Calcular predicciones (usando todo el dataset)
        y_pred = predict(x, w1, w2)

        # b. Calcular y registrar Q(w)
        cost = compute_cost(y_pred, y)
        history['cost'].append(cost)
        history['w1'].append(w1)
        history['w2'].append(w2)

        # c. Calcular gradiente (usando todo el dataset)
        grad_w1, grad_w2 = compute_gradients_batch(x, y, y_pred)

        # d. Actualizar parámetros
        w1 = w1 - eta * grad_w1
        w2 = w2 - eta * grad_w2

        # Imprimir progreso cada 100 iteraciones
        if (i+1) % 100 == 0:
            print(f"Iteración {i+1}/{num_iterations} - Pérdida (Q(w)): {cost:.4f}, w1: {w1:.4f}, w2: {w2:.4f}")

    end_time = time.time()
    print(f"Entrenamiento GD completado en {end_time - start_time:.4f} segundos.")
    return w1, w2, history

# === PARTE 4: Descenso por Gradiente Estocástico (SGD) ===

# 4.1 Implementación del Algoritmo SGD
def stochastic_gradient_descent(x, y, eta, num_epochs):
    # 1. Inicialización de parámetros
    w1 = 0.0
    w2 = 0.0
    N = len(x)
    # Registraremos el coste al final de cada época
    history = {'cost': [], 'w1': [], 'w2': []}
    # Registraremos cada actualización individual para visualizar la trayectoria (history_fine)
    history_fine = {'w1': [0.0], 'w2': [0.0]}

    print("\n--- Iniciando Stochastic Gradient Descent (SGD) ---")
    start_time = time.time()

    # 2. Ciclo de optimización (Épocas)
    for epoch in range(num_epochs):

        # a. Permutar los datos aleatoriamente
        indices = np.random.permutation(N)

        # b. Iterar a través de cada punto
        for i in indices:
            xi = x[i]
            yi = y[i]

            # i. Calcular predicción y error para el punto i
            y_pred_i = predict(xi, w1, w2)
            error_i = y_pred_i - yi

            # ii. Calcular gradientes estocásticos (Gradiente de Q_i)
            grad_w1_i = 2 * error_i
            grad_w2_i = 2 * error_i * xi

            # iii. Actualizar parámetros
            w1 = w1 - eta * grad_w1_i
            w2 = w2 - eta * grad_w2_i

            history_fine['w1'].append(w1)
            history_fine['w2'].append(w2)

        # c. Calcular la pérdida Q(w) al final de la época (para monitoreo)
        # Nota: Calculamos la pérdida sobre todo el dataset para ver la convergencia real.
        y_pred_epoch = predict(x, w1, w2)
        cost_epoch = compute_cost(y_pred_epoch, y)
        history['cost'].append(cost_epoch)
        history['w1'].append(w1)
        history['w2'].append(w2)

        if (epoch+1) % 10 == 0:
             print(f"Época {epoch+1}/{num_epochs} - Pérdida (Q(w)): {cost_epoch:.4f}, w1: {w1:.4f}, w2: {w2:.4f}")

    end_time = time.time()
    print(f"Entrenamiento SGD completado en {end_time - start_time:.4f} segundos.")
    return w1, w2, history, history_fine
````
