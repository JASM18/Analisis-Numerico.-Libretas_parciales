# **A.4.6 Python Programming Basics - Functions** 📌

## **Exercice A.12** 📑

Define the function $g(n)=n^2+n+41$ as a Python function. Write a loop that gives the output for this function for integers from $n=0$ to $n=39$ It is curious to note that each of these outputs is a prime number (check this on your own). Will the function produce a prime for $n=40$? For $n=41$?

### Código A.12 👨‍💻

````python
import math

def g(n):   # La funcion g(n) principal
  return n**2 + n + 41

def es_primo(numero): # Indica si cualquier numero dado es primo o no
  for i in range(2, int(math.sqrt(numero)) + 1): # Busca un divisible entre 2 y su raiz cuadrada
    if numero % i == 0: # Si encuentra un divisible tal que su reciduo es 0, entonces no es primo
      return False
  return True # si no lo encuentra, entonces es un numero primo

def todos_primos(inicio, fin):  # Indica si todos los resultados son primos

    for n in range(inicio, fin+1):
        valor = g(n)
        if not es_primo(valor):
            return False
    return True

for j in range(0,40) :
  print("g(",j,") =", g(j))

print("\n¿Todos son primos en [0,39]? ->", todos_primos(0, 39))

print("\n¿Todos son primos en [0,40]? ->", todos_primos(0, 40))
print("g(40) =", g(40), "¿Es primo? ->" ,es_primo(g(40)))

print("\n¿Todos son primos en [0,41]? ->", todos_primos(0, 41))
print("g(41) =", g(41), "¿Es primo? ->" ,es_primo(g(41)))

# Para g(40) y g(41) los resultados que da la funcion NO son numeros primos, pero de 0 al 39 todos los resultados que arroja g(n) para numeros enteros son primos.
````

## **Exercice A.13** 📑

The previous example is a special case of the Babylonian Algorithm for calculating square roots. If you want the square root of $S$ then iterate the sequence

$\quad$ $x_{n+1}=\frac{1}{2}\left(x_n+\frac{S}{x_n}\right)$

until you are within an appropriate tolerance.

Modify the code given in the previous example to give a list of approximations of the square roots of the natural numbers 2 through 20, each to within $10^{-8}$ This problem will require that you build a function, write a ‘for’ loop (for the integers 2-20), and write a ‘while’ loop inside your ‘for’ loop to do the iterations.

### Código A.13 👨‍💻

````python
from math import sqrt

def f(x,i): # donde i es el numero que queremos aproximar su raiz
    return(0.5*(x + i/x))

for i in range(2,21) :
  x = 1.1 # arbitrary starting point
  while abs(x-sqrt(i)) > 10**(-8):
      x = f(x,i)
  print("\nS =", i)
  print("aproximacion:", x)
  print("Raíz exacta: " , sqrt(i))
  print("Error absoluto: " , abs(x - sqrt(i)))
````

# **A.5.1 Numerical Python with `numpy` - Numpy Arrays, Array Operations, and Matrix Operations** 📌

## **Exercice A.17** 📑

Now that we can do some basic linear algebra with numpy it is your turn. Define the matrix $B$ and the vector $u$ as

$B=\left(\begin{array}{ccc}1 & 4 & 8 \\ 2 & 3 & -1 \\ 0 & 9 & -3\end{array}\right) \quad$ and $\quad u=\left(\begin{array}{c}6 \\ 3 \\ -7\end{array}\right)$

Then find

a)$\quad$$Bu$

b)$\quad$$B^2$

c)$\quad$The size and shape of $B$

d)$\quad$$B^Tu$

e)$\quad$The element-by-element product of $B$ with itself

f)$\quad$The dot product of $u$ with the first row of $B$

### `Sintaxis` y $AlgebraLineal$ en *Python* 🔢 (Nota personal) ✏️

---

#### ¿Que es `shape` y `size`? 🤔
- `shape` nos indica la cantidad de **filas** y **columnas** que tiene una matriz.
Por ejemplo, para la matriz $B$ dada anteriormente, su `shape` sería $(3,3)$ (o puedo verlo como la escala que tiene la matriz)

- `size` nos indica la **cantidad** de **elementos** que tiene la matriz. por ejemplo el `size` de $B$ sería $9$ y el `size` de $u$ sería $3$

---

#### ¿Algebra lineal en Python? 🔢
Si taro de hacer una multiplicacion de matrices digamos `B*B` python lo que hará es multiplicar cada $a_{ij}$ por $b_{ij}$, masomenos usa la regla de como se comportan las sumas.

Esto solo sucede si el arreglo esta definido como `np.array`, aun que existen dos maneras de solucionar esto y que la multiplicacion matricial se procese como corresponde según la algebra lineal:

1.   Declarar el arreglo matricial como `np.matrix`, esto va a tratar el arreglo tal cual como se comportaria una matriz matematica, siguiendo asi las reglas de la algebra lineal.
2.   Usar `B@B` en lugar de `B*B`, esto nos permite hacer una multiplicación matricial como corresponde sin la necesidad de declararlo como una matriz, puede seguir siendo un arreglo deffinido como `sp.array`.

En este problema usare la segunda opción, declarando los arreglos como `sp.array` ya que los incisos *e* y *f* es mas sencillo solucionarlo si tratamos nuestras matrices de la forma ya antes mencionada.


### Código A.17 👨‍💻

````python
import numpy as np

B = np.array([[1,4,8],[2,3,-1],[0,9,-3]])
u = np.array([[6],[3],[-7]])

print("a) B*u\n", B @ u ,"\n")
print("b) B^2\n", B @ B ,"\n")
print("c) size y shape\nShape:", B.shape,"\nSize:", B.size,"\n")
print("d) B^T*u\n", B.T @ u, "\n")
print("e) The element-by-eleemnt product of B with itself\n" , B*B , "\n")
print("f) The dot product of u with the first row of B\n" , np.dot(u.ravel(), B[0]))
````

# **A.5.2 Numerical Python with `numpy` - `arrange`, `linspace`, `zeros`, `ones`, and `meshgrid`** 📌

## **Exercice A.18** 📑

Now time to practice with some of these `numpy` commands.

a) Create a `numpy` array of the numbers 1 through 10 and square every entry in the list without using a loop.

b) Create $10\times10$ identity matrix and change the top right corner to a 5. Hint: `np.identity()`

c) Find the matrix-vector product of the answer to part (a) (as a column) and the answer to part (b).

d) Change the bottom row of your matrix from part (b) to all 3’s, then change the third column to all 7’s, and then find the $5^{th}$ power of this matrix.

### Código A.18 👨‍💻

````python
import numpy as np
np.set_printoptions(suppress=True, linewidth=200, formatter={'int': '{:d}'.format}) # Estea linea de codigo sirve para que no me imprima las entradas de la ultima matriz con notacion cientifica, ya que sin esta linea de codigo se veria muy feo visualmente

# --- a) ---
arr = np.arange(1,11) # Creamos un arreglo definido como [1 , 2 , 3 , ... , 10]
print("a)\n",arr**2) # Cada elemento lo elevamos al cuadrado

# --- b) ---
arrIdentity = np.identity(10) # Matriz identidad de 10x10
arrIdentity[0,9] = 5 # Le asignamos un valor 5 a la entrada (1,10) (la esquina superior derecha)
print("\nb)\n",arrIdentity)

# --- c) ---
arrButItIsAColumn = (arr**2).reshape(10,1) # Convertimos el vector del inciso "a" de 1x10 a 10x1 para realizar la multiplicacion correctamente
print("\nc)\n", arrIdentity @ arrButItIsAColumn)

# --- d) ---
arrIdentity[-1,:] = 3 # Llenamos la ultima fila con puros "3"
arrIdentity[:,2] = 7 # Llenamos la 3ra columna de puros 7
print("\nd)\n", arrIdentity @ arrIdentity @ arrIdentity @ arrIdentity @ arrIdentity) # Lo elevamos a 5, osea que se multiplica por si mismo 5 veces
````

## **Exercice A.23** 📑

Define the function $f(x)=3 x^2+x \sin \left(x^2\right)$ symbolically and then do the following:

a)$\quad$Evaluate the function at $x=2$ and get symbolic and numerical answers.

b)$\quad$Take the first and second derivative

c)$\quad$Take the antiderivative

d)$\quad$Find the definite integral from 0 to 1

e)$\quad$Find the limit as $x$ goes to 3

### Comando `display` 📟 (Nota personales) ✏️

Una manera de imprimir con formato $LaTeX$ y/o de forma bonita es usar el comando `display`

Ejemplo en codigo:

```
import sympy as sp
x = sp.Symbol('x')
f = x**2
display(f)
```
Tambien puedo meter algunas operaciones adentro de `display()`

Por ejemplo: `display(sp.diff(f))`






### Código A.23 👨‍💻

````python
import sympy as sp
x = sp.Symbol('x')
f = 3*x**2 + x*sp.sin(x**2)

# --- a) ---
print("a)")
print("\nf(2) =" , f.subs(x,2)) # f.subs(x,2) Sustituye 2 en la variable x
print("f(2) =" , sp.N(f.subs(x,2))) # Obtenemos el valor numerico de la sustitución

# --- b) ---
print("\n\nb)\n")
print("f'(x): ")
display(sp.diff(f,x,1)) # La derivada de una funcion es con sp.diff(funcion , variable a derivar , la "1ra" derivada)
print("\nf''(x): ")
display(sp.simplify(sp.diff(f,x,2))) # sp.diff(funcion , variable a derivar , la "2da" derivada)

# --- c) ---
print("\n\nc)")
print("La integral es: ")
display(sp.integrate(f,x)) # sp.integrate(la funcion , variable a integrar)

# --- d) ---
print("\n\nd)")
print("La integral definida [0,1] es:")
display(sp.integrate(f,(x,0,1))) # sp.integarte(la funcion a integrar, (variable , inicio , fin) )

# --- e) ---
print("\n\ne)")
display(sp.limit(f,x,3)) # sp.limit(la función , la variable , el limite)
print(sp.N(sp.limit(f,x,3))) # valor numerico para el limite
````

# **A.7.5 Symbolic Python with `sympy` - Solving Equations Symbolically** 📌

## **Exercice A.24** 📑

Give the exact and floating point solutions to the equation $x^4-x^2-x+5=0$

### Código A.24 👨‍💻

````python
import sympy as sp

x = sp.Symbol('x')
soln = sp.solve( x**4 - x**2 - x + 5, x)

for j in range(0, len(soln)): # Las soluciones se guardan en un arreglo, asi que tenemos que imprimirla con un for...
    print(sp.N(soln[j]))
````

## **Exercice A.25** 📑

Solve the equation

$\quad$$x^3 \ln (x)=7$

and give your answer both symbolically and numerically.

### Código A.25 👨‍💻

````python
import sympy as sp
from sympy import latex # Para que se escriba bonito
from IPython.display import display, Math # Para que se escriba bonito

x = sp.Symbol('x', positive=True)  # Definido para x > 0

# --- Solución simbólica ---
Simbolic = sp.solve(x**3 * sp.log(x) - 7, x)
print("Solución simbólica:", Simbolic)

display( Math("x = " + latex(Simbolic[0])) ) # Mostrar en latex

# --- Solución numérica ---
Numeric = sp.nsolve(x**3 * sp.log(x) - 7, x, 1)
print("\nSolución numérica:", Numeric)

display(Math("x \\approx " + latex(Numeric.evalf()))) # Mostrar en latex
````

# **A.7.6 Symbolic Python with `sympy` - Symbolic Plotting** 📌

## **Exercice A.26** 📑

Make a symbolic plot of the function $f(x)=x^3 \ln (x)-7$ on the domain $[0,3]$.

### Código A.26 👨‍💻

````python
import sympy as sp
x = sp.Symbol('x')
f = x**3 * sp.log(x) - 7
sp.plot(f,(x,0.01,3)) # No estamos poniendo exactamente [0,3] por que como ln(x) no esta definido en 0 se puede poner extraño
````

## 🔣 `import sympy` **(Algebra, cálculo & Taylor series)**

Esta libreria sirve para trabajar con algebra o calculo, como simplificar, factorizar, derivar o integrar

- `sp.expand(funcion)`: Expandir una ecuación factorizada
```
import sympy as sp
x = sp.Symbol('x')
f = (x+2)**3
sp.expand(f) # do the multiplication to expand the polynomial
```

- `sp.factor(funcion)`: Factorizar
```
import sympy as sp
x = sp.Symbol('x')
h = x**2 + 4*x + 3
sp.factor(h)
```

- Usar identididaes trigonometricas para expandir una expresión
```
import sympy as sp
a, b = sp.symbols('a b')
j = sp.sin(a+b)
sp.expand(j, trig=True)
```

- `sp.simplify(funcion)`: Simplificar expresiones
```
import sympy as sp
x = sp.Symbol('x')
g = x**3 + 5*x**3 + 12*x**2 + 1
sp.simplify(g)
```

- `sp.solve(funcion, variable)`:Encontrar las soluciónes $f(x)=0$
```
import sympy as sp
x = sp.Symbol('x')
h = x**2 + 4*x + 3
sp.solve(h,x)
```

- `sp.subs(función, punto)` & `sp.N`: Evaluar cualquier $f(x)$ en cualquier punto, por ejemplo $f(5)$
```
import sympy as sp
x = sp.Symbol('x')
f = (x+2)**3
f.subs(x,5) # Obtener el valor simbolico
sp.N(f.subs(x,5)) # Obtener el valor númerico
```

- `sp.diff(funcion, variable)`: Derivación $ \frac{d}{dx} f(x) $
```
import sympy as sp
x = sp.Symbol('x') # Define the symbol x
f = (x+2)**3 # Define a symbolic function f(x) = (x+2)^3
df = sp.diff(f,x) # Take the derivative of f and call it "df"
print("f(x) = ", f)
print("f'(x) = ",df)
print("f'(x) = ", sp.expand(df))
```

- `sp.diff(función, variable, n-ecima derivada)`: Calcular la $n$-ecima derivada $ \frac{d^{n}}{dx^{n}} f(x) $
```
mport sympy as sp
x = sp.Symbol('x') # Define the symbol x
f = (x+2)**3 # Define a symbolic function f(x) = (x+2)^3
df = sp.diff(f,x,1) # first derivative
ddf = sp.diff(f,x,2) # second deriative
dddf = sp.diff(f,x,3) # third deriative
ddddf = sp.diff(f,x,4) # fourth deriative
print("f'(x) = ",df)
print("f''(x) = ",sp.simplify(ddf))
print("f'''(x) = ",sp.simplify(dddf))
print("f''''(x) = ",sp.simplify(ddddf))
```

- Derivadas parciales y funciones de varias variables $ \frac{\partial}{\partial x} f(x,y) $
```
import sympy as sp
x, y = sp.symbols('x y') # Define the symbols
f = sp.sin(x*y) + sp.cos(x**2) # Define the function
fx = sp.diff(f,x)
fy = sp.diff(f,y)
print("f(x,y) = ", f)
print("f_x(x,y) = ", fx)
print("f_y(x,y) = ", fy)

```

