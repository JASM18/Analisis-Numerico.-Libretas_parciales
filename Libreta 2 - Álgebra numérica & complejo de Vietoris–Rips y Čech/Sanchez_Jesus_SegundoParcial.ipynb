### **Método de Bisección** 👨‍💻 (implementación)

````python
import sympy as sp

x = sp.Symbol('x')

def Bisection(f, a, b, tol=None):
  # f: funcion de python o lambda
  # a: limite inferior del intervalo
  # b: limite superior del intervalo
  # tol: tolerancia para |f(m)| < epsilon -> donde epsilon es la tolerancia

  if tol is None: # si no se agrego tolerancia, por default es 1x10^-8
    tol = 1e-8

  # Aseguramos que a < b
  if a > b:
    a, b = b, a

  # Checamos si hay cambio de signo
  if sp.N(f.subs(x, a)) * sp.N(f.subs(x, b)) > 0:
    return "No hay garantía de que exista una raíz en ese intervalo"

  # Inicializamos
  m = (a + b) / 2
  iter_contador = 0
  max_iter=1000

  while abs(f.subs(x, m)) > tol and iter_contador < max_iter:
    m = (a + b) / 2

    # Evaluamos extremos y punto medio
    fa = sp.N(f.subs(x, a))
    fb = sp.N(f.subs(x, b))
    fm = sp.N(f.subs(x, m))

    if fa * fm < 0: # si f(a)*f(m) es negativo -> [a,m]
          b = m
    elif fb * fm < 0: # si f(b)*f(m) es negativo -> [m,b]
      a = m
    elif fa == 0: # si f(a)=0  ->  x^* := a
      return a
    elif fb == 0: # si f(b)=0  ->  x^* := b
      return b
    elif fm == 0: # si f(m)=0  ->  x^* := m
      return m

    iter_contador += 1
  print("contador: ", iter_contador)
  return m

f = x**2 - 3*x +1
raiz = Bisection(f, 0, 2, tol=1e-8)
print("Raíz aproximada:", raiz)
print("Valor de la función en la raíz:", f.subs(x, raiz))
````

# **Método de Falsi** 📌

Este método es similar al método de bisección, pero en vez de tomar el punto medio $m$, ahora calculamos la interseccón de la recta secante entre un punto $(a, f(a))$ y otro punto $(b, f(b))$

así, el nuevo punto $m$ se obtiene:
\begin{align}
m = \frac{a \cdot f(b) - b \cdot f(a)}{f(b) - f(a)}
\end{align}

A cotinuación, se muestra graficamente como es que esta formado $m$ con este método,

→ **Algoritmo de Falsi**

1.   verificar si $f(a)$ y $f(b)$ tienen signos opuestos
2.   Si tienen diferentes signos, podemos seguir con el algoritmo y calculamos $m$:

\begin{align}
m = \frac{a \cdot f(b) - b \cdot f(a)}{f(b) - f(a)}
\end{align}

3.   Evaluamos

$\quad\quad$ - Si $f(m)=0 \text{ OR } |f(m)| < ε$, entonces $x^*=m$

$\quad\quad$ - Si $f(a) \cdot f(m) < 0$, la raíz está en $[a,m]$ → $b := m$

$\quad\quad$ - Si $f(m) \cdot f(b) < 0$, la raíz está en $[m,b]$ → $a := m$

4.   Repetir hasta cumplir la tolerancia o un máximo de iteraciones

A continuacón, un ejemplo gráfico de como es que se genera el punto $m$:

## **Método de Falsi** 👨‍💻 (implementación)

````python
import sympy as sp

x = sp.Symbol('x')

def ReglaFalsi(f, a, b, tol=None):
  if tol is None:
    tol = 1e-8

  if a > b:
    a, b = b, a

  fa = sp.N(f.subs(x, a))
  fb = sp.N(f.subs(x, b))

  if fa * fb > 0:
    return "No hay garantía de que exista una raíz en ese intervalo"

  m = a
  iter_contador = 0
  max_iter = 1000

  while abs(f.subs(x, m)) > tol and iter_contador < max_iter:
    # Fórmula Regula Falsi
    m = (a*fb - b*fa) / (fb - fa)
    fm = sp.N(f.subs(x, m))

    if fa * fm < 0:
      b, fb = m, fm   # actualizamos b y fb
    elif fb * fm < 0:
      a, fa = m, fm   # actualizamos a y fa
    elif fa == 0:
      return a        # Si de casualidad f(a) == 0
    elif fb == 0:
      return b        # Si de casualidad f(b) == 0
    elif fm == 0:
      return m        # Si de casualidad f(m) == 0

    iter_contador += 1
  print("contador: ", iter_contador)
  return m

# Ejemplo de uso
f = x**2 - 3*x + 1
raiz = ReglaFalsi(f, 0, 2, tol=1e-8)
print("Raíz aproximada:", raiz)
print("Valor de la función en la raíz:", f.subs(x, raiz))
````

# **Método de Newton-Raphson** 📌

El método de Newton-Raphson (N-R) funciona principalmente gracias a la derivada, en dondevamos comenzamos con un valor inicial arbitrario $x_0$ y se itera el algoritmo hasta que converga hacia $x^*$. Vamos a empezar definiendo el algoritmo:

Sea $x_0$ un valor en el eje $x$ y $x^*$ es la raíz de la funcón.


1.   Escoger un valor arbitrario para $x_0$
2.   Si $f(x_0)=0$, entonces $x^*:=x_0$ y se acabo.
3.   Si no ($f(x_0) \not= x^* $), entonces construimos la recta tangente en el punto $x_0, f(x_0)$ en donde $m$ es la pendiente de dicha recta:

\begin{align}
\left.
\begin{array}{l}
m = f'(x_0) \\
\text{punto: }(x_0, f(x_0))
\quad\end{array}
\right\}
&\quad y - f(x_0) = f'(x_0)\,(x-x_0)
\end{align}


Ahora el chiste de este método es encontrar $x_1$ que sera nuestro siguiente candidadto para que sea $x^*$, luego encontrar $x_2$ y asi hasta un tal $x_{n+1}$ que sea lo suficientemente cercano a $x^*$.

5.   Ya tenemos la recta tangente para el punto $x_0$, ahora hay que calcular la raíz de esa recta que sera nuestra nueva candidata $x_1$

$\quad\quad \text{para sacar la raíz, tomamos como } y=0 \text{ y lo sistituimos en la ecuación de la recta}$
\begin{align}
y = 0 : -f(x_0) = f'(x_0) (x_1-x_0)
\end{align}

$\qquad \text{Ahora despejamos } x_1$

\begin{align}
-f(x_0) = f'(x_0) (x_1-x_0) \quad ⇒ \quad \frac{-f(x_0)}{f'(x_0)} = x_1 - x_0
\end{align}

\begin{align}
x_1 = x_0 - \frac{f(x_0)}{f'(x_0)}
\end{align}

Ya hemos encontrado $x_1$, así que ahora simplemente en la siguiente iteración trabajaremos con $x_1$ para encontrar a $x_2$ y asi sucesivamente hasta aproximarnos a $x^*$

De forma generalizada:

\begin{align}
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
\end{align}


Es de esperar que este método tiene algunos problemas, alguno de ellos son:


*   No todas las funciones son diferenciables en todo su intervalo
*   Si de casualidad, resulta que $f'(x_n) = 0$, entonces no habrá ninguna raíz para esa recta tangente, por lo tanto tampoco existirá otro sucesor $x_{n+1}$
*   Dependiendo del valor que le demos a $x_0$, existe la posibilidad de que el algoritmo no converga hacía $x^*$.




## **Método de Newton-Raphson** 👨‍💻 (implementación)

````python
import sympy as sp

x = sp.Symbol('x')

def NewtonRaphson(f, x0, tol=None):
    if tol is None:
        tol = 1e-8

    # Derivada simbólica
    fprima = sp.diff(f, x)

    # Inicialisamos
    xn = x0
    iter_contador = 0
    max_iter=100

    while abs(f.subs(x, xn)) > tol and iter_contador < max_iter :
        fxn = sp.N(f.subs(x, xn))
        fpxn = sp.N(fprima.subs(x, xn))

        if fpxn == 0:
            return "FALLO f'(x_n) = 0, fallo en la iteración: ", iter_contador

        # Fórmula Newton-Raphson
        xnext = xn - fxn/fpxn

        xn = xnext
        iter_contador += 1
    print("contador: ", iter_contador)
    return xn

# Ejemplo de uso
f = x**2 - 3*x + 1
raiz = NewtonRaphson(f, 0, tol=1e-8)
print("Raíz aproximada:", raiz)
print("Valor de la función en la raíz:", f.subs(x, raiz))
````

# **Método de la secante** 📌

El método de la secante es muy similar al método de Newton-Raphson, pero ahora en vez de usar la recta tangente, aproximamos dicha recta pero ahora a travez de la recta secante.

Ejemplo visual de la idea:

Primero hay que recordar que la pendiente esta definida como:

\begin{align}
m = \frac{f(x_1) - f(x_0)}{x_1 - x_0}
\end{align}

Mientras que la ecuación del método de $\text{N-R}$ es:
\begin{align}
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
\end{align}

Entonces, la formula para este neuvo método es muy similar, solo hay que remplazar, de la formula, la pendiente $f'(x_n)$ por la pendiente $m = \frac{f(x_1) - f(x_0)}{x_1 - x_0}$:

\begin{align}
x_{n+1} = x_n - \frac{f(x_n)}{\frac{f(x_1) - f(x_0)}{x_1 - x_0}} \quad ⇒ \quad x_{n+1} = x_n - \frac{f(x_n) \cdot (x_n - x_{n-1})}{f(x_n)-f(x_{n-1})}
\end{align}

Así, dado $x_0$ y $x_1$, se itera para calcular $x_3$, $x_4$ y así hasta encontrar $x^*$

Este método arregla el problema de derivación de $\text{N-R}$, sin embargo sigue presentando sus mismos problemas:
*   Si de casualidad, resulta que $f(x_n)-f(x_n-1)= 0$, entonces no habrá ninguna raíz para esa recta secante, por lo tanto tampoco existirá otro sucesor $x_{n+1}$
*   Dependiendo del valor que le demos a $x_0$ y $x_1$, existe la posibilidad de que el algoritmo no converga hacía $x^*$.

````python
import sympy as sp

x = sp.Symbol('x')

def Secante(f, x0, x1, tol=None):
    if tol is None:
        tol = 1e-8

    # Inicialisamos
    x_n1 = x0
    x_n = x1

    iter_contador = 0
    max_iter=100

    while abs(f.subs(x, x_n)) > tol and iter_contador < max_iter :
        denominador = sp.N(f.subs(x, x_n) - f.subs(x, x_n1))
        nominador = sp.N(f.subs(x, x_n) * (x_n - x_n1))

        if denominador == 0:
            return "FALLO Divición entre 0, fallo en la iteración: ", iter_contador

        # Fórmula secante
        xnext = x_n - nominador / denominador

        x_n1 = x_n
        x_n = xnext

        iter_contador += 1
    print("contador: ", iter_contador)
    return x_n

# Ejemplo de uso
f = x**2 - 3*x + 1
raiz = Secante(f, 0, 0.5, tol=1e-8)
print("Raíz aproximada:", raiz)
print("Valor de la función en la raíz:", f.subs(x, raiz))
````

# **Método del punto fijo** 📌

El objetivo de este método pues como los otros, es encontrar $f(x)=0$, el chiste de este método es re escribir la funcíon de la forma $x = g(x)$ y luego usar la iteracíon $x_{n+1} = g(x_n)$

Pasos:
*   Partimos de una suposición inicial $x_0$
*   Calculamos $x_1 = g(x_0)$
*   Luego $x_2 = g(x_1)$ y así sucesivamente...
*   El proceso se repite hasta que la diferencia entre dos iteraciones consecutivas sea mas pequeña que la tolerancia:

\begin{align}
|x_{n+1} - x_n| < ε
\end{align}

El método no siempre converge, para que lo haga, debe de cumplirse:

\begin{align}
|g'(x)| < 1 \quad \text{en la vecindad de la raíz}
\end{align}

Tomemos de ejemplo

\begin{align}
f(x) = x^2 - 3x + 1 \quad ⇒ \quad 0 = (x^2 + 1) - 3x \quad ⇒ \quad 3x = x^2+1
\end{align}

\begin{align}
x = \frac{x^2+1}{3} \quad ⇒ \quad g(x) = \frac{x^2+1}{3}
\end{align}

Asi, si empezamos con $x_0 = 0.5$, $x_1 = g(0.5)$ y así hasta $x_{n+1} = g(x_n)$

````python
import sympy as sp

x = sp.Symbol('x')

def PuntoFijo(g, x0, tol=None, max_iter=1000):
    if tol is None:
        tol = 1e-8

    xn = x0
    iter_contador = 0

    while iter_contador < max_iter:
        xnext = sp.N(g.subs(x, xn))  # Evaluamos g(xn)

        if abs(xnext - xn) < tol:    # Criterio de paro
            print("contador:", iter_contador)
            return xnext

        xn = xnext
        iter_contador += 1

    return "⚠️ No convergió en el máximo de iteraciones"

# Ejemplo de uso
f = x**2 - 3*x + 1
# Reescribimos f(x)=0 como x = (x^2+1)/3
g = (x**2 + 1)/3

raiz = PuntoFijo(g, 0.5, tol=1e-8)
print("Raíz aproximada:", raiz)
print("Valor de f en la raíz:", f.subs(x, raiz))
````

# **Comparación de los diferentes métodos númericos** 📌

En este apartado, se va a comparar los 5 métodos para aproximar la raíz, a continuación se va a presentar el **orden de convergencia** que describe que tan rápido los algoritmos se acerca la aproximación $x_n$ a la raíz verdadera $x^*$ en cada iteración.

**Órdenes Teóricos de Convergencia:**
* **Lineal ($O(h)$):** El error se reduce por un factor aproximadamente constante en cada paso ($E_{n+1} \approx C \cdot E_n$). Se espera este comportamiento de **Bisección**, **Regla Falsi** y **Punto Fijo** (bajo condiciones apropiadas para $g'(x)$).
* **Superlineal ($O(h^{1.618})$):** La convergencia es más rápida que lineal. Es el orden característico del método de la **Secante**.
* **Cuadrática ($O(h^2)$):** El error se reduce aproximadamente con el cuadrado del error anterior ($E_{n+1} \approx C \cdot E_n^2$). Es el orden más rápido y es característico del método de **Newton-Raphson**.

#### Demostración Empírica

Para demostrar empíricamente estos órdenes, seleccionamos una función de prueba $f(x)$ con una raíz exacta conocida $x^*$. Para cada método, calculamos el error absoluto $E_n = |x_n - x^*|$ en cada iteración $n$ y lo graficamos.

Al usar una escala logarítmica en el eje Y (Error), se espera que:
* La convergencia **lineal** aparezca como una **línea recta descendente**.
* La convergencia **cuadrática** aparezca como una **curva que cae drásticamente**, volviéndose casi vertical.

**Función de Prueba:**
$$f(x) = x^2 - 3x + 1$$
**Raíz Exacta:**
$$x^* = \frac{3 + \sqrt{5}}{2} \approx 2.61803398...$$
**Intervalo de Búsqueda (Bisección/Falsi):** $[2, 3]$ (ya que $f(2) = -1$ y $f(3) = 1$)
**Puntos Iniciales (Secante/Newton/P.Fijo):** $x_0 = 2.0$, $x_1 = 3.0$

A continuación, se presentan las gráficas de convergencia de los cinco métodos aplicados a esta función:

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# ======================================================================
# --- Parte 1: Gráfico de Convergencia Empírica (Combinado) ---
# ======================================================================
print("--- Parte 1: Gráfico de Convergencia Empírica (Combinado) ---")

# --- 1. Función de Prueba y Raíz Exacta ---
f = lambda x: x**2 - 3*x + 1
f_prime = lambda x: 2*x - 3 # Derivada para Newton
# Raíz exacta (cuadrática): (3 + sqrt(9-4))/2 = (3 + sqrt(5))/2
x_exacto = (3 + np.sqrt(5)) / 2
# g(x) para Punto Fijo: g(x) = x - 0.1 * f(x)
g_fijo = lambda x: x - 0.1 * (x**2 - 3*x + 1)

print(f"Función de prueba: f(x) = x^2 - 3x + 1")
print(f"Raíz exacta: x* = (3 + sqrt(5))/2 ≈ {x_exacto:.12f}")

# --- 2. Métodos de Búsqueda de Raíces (Modificados para devolver historial de error) ---
# (Se asume que las funciones Bisection_list, ReglaFalsi_list, Secante_list,
# Newton_list, y PuntoFijo_list están definidas como en el bloque anterior)

def Bisection_list(f, a, b, tol=1e-8, max_iter=100):
    if a > b: a, b = b, a
    fa = f(a); fb = f(b)
    if np.sign(fa) * np.sign(fb) >= 0:
        print("Bisección: Mismos signos.")
        return [], []
    errors = []; iterations = []; iter_count = 0
    m = a + (b - a) / 2
    errors.append(abs(m - x_exacto)); iterations.append(iter_count)
    while iter_count < max_iter:
        iter_count += 1
        m = a + (b - a) / 2; fm = f(m)
        errors.append(abs(m - x_exacto)); iterations.append(iter_count)
        if (b - a) / 2 < tol or abs(fm) < tol: return iterations, errors
        if np.sign(fm) == np.sign(fa): a = m; fa = fm
        else: b = m
    return iterations, errors

def ReglaFalsi_list(f, a, b, tol=1e-8, max_iter=100):
    if a > b: a, b = b, a
    fa = f(a); fb = f(b)
    if np.sign(fa) * np.sign(fb) >= 0:
        print("F. Posición: Mismos signos.")
        return [], []
    errors = []; iterations = []; iter_count = 0
    m = a; m_prev = b
    errors.append(abs(m - x_exacto)); iterations.append(iter_count)
    while iter_count < max_iter:
        iter_count += 1
        try:
            denominador = fb - fa
            if abs(denominador) < 1e-15: raise ZeroDivisionError
            m_prev = m; m = (a*fb - b*fa) / denominador
            if m < min(a,b) or m > max(a,b): m = a + (b-a)/2
        except (ZeroDivisionError, OverflowError): m = a + (b-a)/2
        fm = f(m)
        errors.append(abs(m - x_exacto)); iterations.append(iter_count)
        if abs(m - m_prev) < tol or abs(fm) < tol: return iterations, errors
        if np.sign(fm) == np.sign(fa): a = m; fa = fm
        else: b = m; fb = fm
    return iterations, errors

def Secante_list(f, x0, x1, tol=1e-8, max_iter=100):
    fx0 = f(x0); fx1 = f(x1)
    errors = [abs(x0 - x_exacto), abs(x1 - x_exacto)]
    iterations = [0, 1]; iter_count = 1
    while iter_count < max_iter:
        iter_count += 1
        denominador = fx1 - fx0
        if abs(denominador) < 1e-15: return iterations, errors
        try: x_next = x1 - fx1 * (x1 - x0) / denominador
        except (OverflowError, ZeroDivisionError): return iterations, errors
        fx_next = f(x_next)
        errors.append(abs(x_next - x_exacto)); iterations.append(iter_count)
        if abs(x_next - x1) < tol or abs(fx_next) < tol: return iterations, errors
        x0, fx0 = x1, fx1
        x1, fx1 = x_next, fx_next
    return iterations, errors

def Newton_list(f, f_prime, x0, tol=1e-8, max_iter=100):
    xn = x0
    errors = [abs(xn - x_exacto)]; iterations = [0]; iter_count = 0
    while iter_count < max_iter:
        iter_count += 1
        fxn = f(xn); fpxn = f_prime(xn)
        if abs(fpxn) < 1e-15: return iterations, errors
        try: xn_next = xn - fxn / fpxn
        except (OverflowError, ZeroDivisionError): return iterations, errors
        errors.append(abs(xn_next - x_exacto)); iterations.append(iter_count)
        if abs(xn_next - xn) < tol or abs(fxn) < tol: return iterations, errors
        xn = xn_next
    return iterations, errors

def PuntoFijo_list(g, x0, tol=1e-8, max_iter=100):
    xn = x0
    errors = [abs(xn - x_exacto)]; iterations = [0]; iter_count = 0
    while iter_count < max_iter:
        iter_count += 1
        try: xn_next = g(xn)
        except (OverflowError, ZeroDivisionError): return iterations, errors
        errors.append(abs(xn_next - x_exacto)); iterations.append(iter_count)
        if abs(xn_next - xn) < tol: return iterations, errors
        xn = xn_next
    return iterations, errors

# --- 3. Ejecución y Creación de Gráfico Combinado ---

tol = 1e-10
max_it = 30 # Limitar iteraciones para claridad

# Nuevo intervalo [a, b] = [2, 3] (f(2)=-1, f(3)=1)
a_intervalo = 2.0
b_intervalo = 3.0
x0_start = 2.0 # Punto inicial para Newton y Punto Fijo

plt.figure(figsize=(12, 8))

# --- Método 1: Bisección ---
iters_b, errors_b = Bisection_list(f, a_intervalo, b_intervalo, tol, max_it)
if iters_b:
    plt.plot(iters_b, errors_b, 'o-', label='Bisección (Lineal)')

# --- Método 2: Regla Falsi ---
iters_f, errors_f = ReglaFalsi_list(f, a_intervalo, b_intervalo, tol, max_it)
if iters_f:
    plt.plot(iters_f, errors_f, 's-', label='Regla Falsi (Lineal)', color='orange')

# --- Método 3: Secante ---
iters_s, errors_s = Secante_list(f, a_intervalo, b_intervalo, tol, max_it)
if iters_s:
    plt.plot(iters_s, errors_s, '^-', label='Secante (Superlineal)', color='green')

# --- Método 4: Newton-Raphson ---
iters_n, errors_n = Newton_list(f, f_prime, x0_start, tol, max_it)
if iters_n:
    plt.plot(iters_n, errors_n, 'x-', label='Newton (Cuadrática)', color='red')

# --- Método 5: Punto Fijo ---
iters_pf, errors_pf = PuntoFijo_list(g_fijo, x0_start, tol, max_it)
if iters_pf:
    plt.plot(iters_pf, errors_pf, 'd-', label='Punto Fijo (Lineal)', color='purple')

# --- Formato del Gráfico Combinado ---
plt.yscale('log')
plt.title('Comparativa de Convergencia Empírica ($f(x) = x^2 - 3x + 1$)', fontsize=16)
plt.xlabel('Iteración ($n$)')
plt.ylabel('Error Absoluto $|x_n - x*|$ (Escala Logarítmica)')
plt.legend()
plt.grid(True, which="both")
plt.xlim(left=0)
plt.ylim(bottom=1e-12, top=1e1)
````

El gráfico combinado ilustra perfectamente la teoría:

1.  **Convergencia Cuadrática (Newton):** La línea roja (Newton) cae de forma casi vertical. El error se reduce tan drásticamente que alcanza la tolerancia de $10^{-10}$ en solo 4-5 iteraciones, demostrando su potente convergencia cuadrática.
2.  **Convergencia Superlineal (Secante):** La línea verde (Secante) es claramente la segunda más rápida. Su curva es mucho más pronunciada que la de los métodos lineales, acorde con su orden de convergencia $O(h^{1.618})$.
3.  **Convergencia Lineal (Bisección, Falsi, Punto Fijo):** Las líneas azul (Bisección), naranja (Falsa Posición) y morada (Punto Fijo) muestran pendientes casi constantes en la escala logarítmica. Esto confirma visualmente su orden de convergencia lineal, siendo notablemente más lentas que Newton y Secante.

# **Proyecto de discos** 💿

Este proyecto tiene como objetivo que dado un sistema de discos, encontrar un punto en donde todos los discos se intersecten.

Sea $C$ el centro de un circulo y $r$ el radio de dicho circulo, un disco esta formado por:

\begin{align}
D(c,r) \text{ en ℝ}^2
\end{align}

Un sistema de discos esta definido como

\begin{align}
M = {D_i(c_i ; r_i)} \text{ donde es un sistema finito y } r_i > 0
\end{align}

Partimos el problema dado un valor $λ$ que multiplica todos los radios de los circulos del sistema, así todos los discos se van a expandir o reducir dependiendo del valor exacto de $\lambda$

\begin{align}
M_\lambda = D_i(c_i ; \lambda \cdot r_i)
\end{align}

## **El complejo de $\text{Vietoris–Rips}$ ($\text{V-R}$)**

$\text{V-R}$ es un sistema que se le aplica a nuestro sistema de discos, estas son las ideas clave de $\text{V-R}$:
- En nuestro sistema de discos $M_λ$, diremos para algun valor de λ se forma un "simplejo" (arista, triangulo, etc) en el complejo de $\text{V-R}$ si **cada par de discos se intersecta**
- No se requiere que todos los discos del sisrema tengan un punto en comúm, lo que se necesita es que se intersecten por pares
- El complejo de $\text{V-R}$ no esta limitado a solo un valor de $λ$, si no que es a partir de un valor $λ$ que se forma el complejo de $\text{V-R}$ (vamos a denotar dicho lambda como $λ_\text{V-R}$), entonces el complejo de $\text{V-R}$ se forma a partir de algun $ λ_n ≥ λ_\text{V-R}$

El complejo de $\text{V-R}$ nos da una aproximación de cómo se conectan los discos para ir escalando con $λ$, sin exigir que todos compartan un punto en común, solo con que todos los pares se intersecten entre si.

### Por ejemplo:
Si tenemos 3 discos $D_1, D_2, D_3$ y todos los pares se intersectan $(D_1,D_2),(D_1,D_3),(D_2,D_3)$, entonces aparece un triangulo en el complejo $\text{V-R}$, aunque los tres discos no comparten un punto en común

image.png

La ecuación para obtener el valor de $λ$ para un par de discos es:

\begin{align}
λ_{i,j} = \frac{d(c_i, c_j)}{r_i+r_j}
\end{align}

en donde $d(c_i, c_j)$ es la distancia entre los centros

Esa es la ecuación para algun par de discos $(D_i, D_j)$, pero para calcular cual es el $λ$ que forma $\text{V-R}$ hay que tomar el $λ_{i,j}$ mayor que resulte tras escalar todos los pares

\begin{align}
λ_{\text{V-R}} := \text{max } \lambda_{ij}
\end{align}

Muy IMPORTANTE aclarar que el valor que tome $λ_{\text{V-R}}$ representa la **escala mínima** en la que el complejo $\text{V-R}$ conecta a todo el sistema de discos.
- Para algun $λ < λ_{\text{V-R}}$: no todos los pares se intersectan aún
- Para $λ ≥ λ_{\text{V-R}}$: todos los pares ya se intersectan y por lo tanto el complejo $\text{V-R}$ esta completo.


Vale la pena mencionar:
- Si solo se intersectan dos discos, aparece una arista.
- Si tres discos se intersectan par a par, aparece un triangulo (2-simplejo).
- Si cuatro discos se intersectan par a par, aparece un tetraedro (3-simplejo).

Esto subraya que el complejo de $\text{V-R}$ no es solo un dibujo de discos, sino un complejo simplicial
.

Y como un detalle a comentar para mas adelante, el complejo de Cech trata tambien de encontrar un $\lambda _{\text{Cech}}$, solo que dicho complejo es como "mas grande" que $λ_{\text{V-R}}$

\begin{align}
λ_{\text{Cech}} ≥ λ_{\text{V-R}}
\end{align}

Pero eso lo veremos mejor mas adelante, por ahora vamos con ejemplos para el complejo de $\text{V-R}$

### Ejemplo practico

Lo que hay que hacer es primero agarrar $D_1$, luego agarramos una pareja para ese disco $D_2$, calculamos $\lambda _{1,2}$ y pues automaticamente ese será $λ_{\text{V-R}}$, ahora agarramos su siguiente pareja $D_3$ para calcular $\lambda _{1,3}$, ahora comparamos ambos $\lambda$ para saber cual es mayor, si $\lambda _{1,3}$ es mayor, entonces ahora $λ_{\text{V-R}} := \lambda _{1,3}$ y así vamos comparando cada disco con cada disco para determinar el valor de $λ_{\text{V-R}}$

En código, seria algo como

```
# Con esos dos ciclos for, comparamos cada disco (i) con todos los discos (j)

for i in range(len(MatrizDiscos)) : # "i" recorre cada disco
  for j in range(i+1, len(MatrizDiscos)) : # "j" recorre todos los discos

    lambda_ij = distancia(c_i, c_j) / (r_i + r_j) # Calculamos el lambda de la pareja de discos i,j

    if lambda_ij > lambda_vietorisRips :
      lambda_vietorisRips = lambda_ij

```

Cabe aclarar que lo anterior no es un código oficial, es solo un prototipo para dar la idea algoritmica de como encontrar $λ_{\text{V-R}}$

# **El complejo de $\text{Čech}$**

El sistema de $\text{Čech}$ tambien trata de re-escalar un sistema de discos que cumple ciertas condiciones
- Tienes un sistema de discos ya re-escalado $M_\lambda = \{D(c_i, \lambda r_i)\}$
- En el sistema de $\text{Čech}$ aparece  un *simplejo* (arista, triangulo, etc.) si y solo si la **intersección de todos los discos correspondientes es no vacía**

Vamos a desarrollar este ultimo puntto, La diferencia entre este sistema y el sistema de $\text{V-R}$ es que en $\text{Čech}$ NO BASTA que se intersecten de a pares, si no que ahora todos los discos deben de tener un **punto en común**

image.png

Resumidamente:
- $\text{V-R}$: los discos se intersectan por pares.
- $\text{Čech}$: todos los discos se intersectan en un solo punto.


### Ejemplo: caso 2 discos $(D_1, D_2)$
si en el sistema $M$ solo tenemos 2 discos $(D_1, D_2)$, para que se forme el sistema de $\text{Čech}$ basta con que esos dos discos se intersecten, en este ejemplo es el mismo criterio que $\text{V-R}$, así que para 2 discos ambos sistemas coinciden

### Ejemplo: caso 3 discos $(D_1, D_2, D_3)$
En $\text{V-R}$ basta con que cada par se intersecte, pero ahora en $\text{Čech}$ debe exsistir un punto en común con los tres discos

Esto nos indica que un sistema $\text{Čech}$ es también un sistema de $\text{V-R}$, ya que en $\text{Čech}$ de por sí ya todas ya se estaban intersectando. Generalmente, cada sistema de $\text{Čech}$ es también un sistema de $\text{V-R}$, pero la afirmación opuesta no se cumple en general

\begin{align}
\text{Čech} (M_\lambda) ⊆ \text{V-R} (M_\lambda)
\end{align}

El siguiente colorario se saco directamente del articulo que nos mando el profe Espinoza, el cual es muy importante

## **Colorario 1**
Si $M$ es un sistema de disco $d$ (Numero de discos en el sistema) arbitrario y $v_M$ ($\lambda_\text{V-R para nosotros}$) su escala de Vietoris-Rips, entonces su escala $\text{Čech}$ satisface:

\begin{align}
μ_M ∈ [ν_M , \sqrt{ \frac{2d}{d+1} } v_M]
\end{align}
En consecuencia, para cada sistema de disco $d$ $M$, el sistema de disco re-escalado $M_{ \sqrt{2d / (d+1)}_{v_M}}$ es siempre un sistema de $\text{Čech}$


Para resumir lo anterior, vamos a denotar el complejo de $\text{V-R}$ como $V_M$ y el complejo de $\text{Čech}$ como $𝓶_M$. Gracias al colorario que vimos anteriormente, sabemos que nuestro sistema de $\text{Čech}$ que buscamos esta en el intervalo:

\begin{align}
V_M ≤ 𝓶_M ≤ \sqrt{ \frac{2d}{d+1}} \cdot V_M\\\\
\end{align}

Ya planteado lo anterior, ahora ya podemos preguntarnos ¿Como encontrar el punto y el factor de escala $\lambda _\text{Čech}$ minimo para que exista este complejo?

## Intersección de dos discos ($d_{ij}$ y $d_{ji}$)

Cuando los discos re-escalados ($D_i$, $D_j$) se tocan, las fronteras de sus circulos se cruzan en dos puntos. Estos son los puntos $d_{ij}$ y $d_{ji}$. Son los candidatos perfectos para ser el punto de intersección común, por que si un tercer disco $D_k$ puede "alcanzar" uno de ellos, entonces ya se tendria una intersección entre los tres discos

image.png

Entonces regresando a la pregunta, ¿Como calcular $λ_{\check{C}}$?

Aquí es donde entra la función protagonista $ρ_M(λ)$ el cual nos describe el estado de intersección global.

La función $ρ_M(λ)$ esta definida como:

$$
\rho_M(\lambda) := \max_{1 \le i < j \le m} \left\{ \min_{k \ne i,j} \left\{ \lambda r_k - \|d_{ij}(\lambda) - c_k\| \right\} \right\}
$$
Donde:
* $M$ es el sistema de $m$ discos.
* $\lambda$ es el factor de escala que estamos probando.
* $d_{ij}(\lambda)$ es un punto de intersección entre los bordes de los discos escalados $D_i(\lambda)$ y $D_j(\lambda)$.
* El término interno, $\rho_k = \lambda r_k - \|d_{ij}(\lambda) - c_k\|$, mide la "holgura" o distancia con signo del punto $d_{ij}$ al borde del disco $D_k(\lambda)$. Si $\rho_k \ge 0$, el punto $d_{ij}$ está *dentro* del disco $k$

**Significado de $\rho_M(\lambda)$:**
Esta función busca el punto de intersección $d_{ij}$ (generado por cualquier par $i,j$) que esté "más centralmente" ubicado con respecto a *todos los demás* discos $k$.
* Si $\rho_M(\lambda) < 0$: Significa que *ningún* punto de intersección de ningún par está contenido en la intersección de todos los demás discos. Por lo tanto, la intersección global $\bigcap_{i=1}^m D_i(\lambda)$ está vacía.
* Si $\rho_M(\lambda) \ge 0$: Significa que existe *al menos un* punto de intersección $d_{ij}$ que también está contenido en todos los demás discos $k$. Esto implica que la intersección global es no vacía.

El **lambda de Čech ($\lambda_{\check{C}}$)** es el valor exacto de $\lambda$ donde $\rho_M(\lambda)$ cruza de negativo a positivo, es decir, **la raíz de $\rho_M(\lambda) = 0$**.

**2. Búsqueda de la Raíz**

Para encontrar esta raíz, aplicamos los métodos de Bisección, Regla Falsi y Secante. Usamos el intervalo de búsqueda $[\nu_M, \sqrt{4/3}\nu_M]$ provisto por el lema generalizado de Vietoris-Rips, donde $\nu_M = \lambda_{VR}$.

* Se inicia la búsqueda en $a = \nu_M$. Se comprueba si $\rho_M(a) \ge 0$. Si es así, el caso es trivial y $\lambda_{\check{C}} = \nu_M$.
* Si $\rho_M(a) < 0$, se busca la raíz en el intervalo $[a, b]$, donde $b = a \cdot \sqrt{4/3}$ (con ajustes si $\rho_M(b)$ no es positivo).

````python
import time
import pandas as pd # Para guardar y mostrar los resultados

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import itertools
import time

x = sp.Symbol('x')

# ======================================================================
# --- FUNCIONES ---
# ======================================================================

# Esta función crea el sistema de discos original
def IniciarSistemaDiscos(NumDiscos):

  # Para cada disco, se le asigna un centro y un radio, D = { c(x,y) ; r } en donde c(x,y) son valores aleatorios enteros entre -10 y 10 y radio es un aleatorio entre 1 y 3
  cx = np.random.randint(-20, 21, size = NumDiscos)
  cy = np.random.randint(-20, 21, size = NumDiscos)
  r = np.random.uniform(1, 1.5 , size = NumDiscos) # 1, 1

  # Construir la matriz MatrizDiscos
  MatrizDiscos = np.column_stack((cx, cy, r))
  return MatrizDiscos

# ======================================================================

def DibujarDiscos(ax, MatrizDiscos, titulo = "Sistema de discos") :
  NumDiscos = len(MatrizDiscos)
  colors = plt.cm.tab10(np.linspace(0, 1, NumDiscos))

  for (cx, cy, r), color in zip(MatrizDiscos, colors):
    #  Dibuja el círculo de color.
    circle_fill = plt.Circle((cx, cy), r, fill=True, alpha=0.5, color=color,
      label=f"({cx},{cy},r={r:.1f})")
    ax.add_patch(circle_fill)

    # Dibuja un segundo círculo encima del circulo principal para darle un contorno negro
    circle_edge = plt.Circle((cx, cy), r, fill=False, edgecolor='black', linewidth=1.5)
    ax.add_patch(circle_edge)

  # Calcular límites de la cuadricula considerando los radios
  x_min = np.min(MatrizDiscos[:,0] - MatrizDiscos[:,2]) - 1
  x_max = np.max(MatrizDiscos[:,0] + MatrizDiscos[:,2]) + 1
  y_min = np.min(MatrizDiscos[:,1] - MatrizDiscos[:,2]) - 1
  y_max = np.max(MatrizDiscos[:,1] + MatrizDiscos[:,2]) + 1

  ax.set_xlim(x_min, x_max)
  ax.set_ylim(y_min, y_max)
  ax.set_aspect('equal', adjustable='box')
  ax.set_title(titulo)
  ax.grid(True)

  # Ejes y título
  plt.grid()
  plt.title(titulo)
  plt.xlabel("coordenada x")
  plt.ylabel("coordenada y")
  plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")

# ======================================================================

def distancia(punto1, punto2) :
    p1 = np.asarray(punto1); p2 = np.asarray(punto2)
    return np.linalg.norm(p1 - p2)

# ======================================================================

def VietorisRips(MatrizDiscos) :
    NumDiscos = len(MatrizDiscos)
    lambda_vietorisRips = 0
    if NumDiscos < 2: return 0

    for i in range(NumDiscos) :
        for j in range(i+1, NumDiscos) :
            dist_ij = distancia(MatrizDiscos[i][:2], MatrizDiscos[j][:2])
            suma_radios = MatrizDiscos[i][2] + MatrizDiscos[j][2]
            if suma_radios < 1e-12: continue
            lambda_ij = dist_ij / suma_radios
            if lambda_ij > lambda_vietorisRips : lambda_vietorisRips = lambda_ij
    return lambda_vietorisRips

# ======================================================================

def interseccion_dos_discos(disco1, disco2, lambda_escala=1.0):

    c1 = disco1[:2]; r1 = disco1[2] * lambda_escala
    c2 = disco2[:2]; r2 = disco2[2] * lambda_escala

    # Asegurar radios no negativos
    if r1 < 0: r1 = 0
    if r2 < 0: r2 = 0
    d = np.linalg.norm(c1 - c2); epsilon = 1e-9

    # No intersección o contención sin tocarse o centros iguales con radios diferentes
    if d > r1 + r2 + epsilon or d < abs(r1 - r2) - epsilon or (d < epsilon and abs(r1-r2) > epsilon) : return []

    # Centros iguales y radios iguales (mismo disco)
    if d < epsilon and abs(r1-r2) < epsilon: return [] # ¿O devolver algo especial? Devolvemos vacío.

    # Tangencia
    if abs(d - (r1 + r2)) < epsilon or abs(d - abs(r1 - r2)) < epsilon:
        if abs(d) > epsilon : # Evitar división por cero si d=0
            if abs(d - (r1 + r2)) < epsilon: punto_tang = c1 + r1 * (c2 - c1) / d
            else: punto_tang = c1 + r1 * (c2 - c1) / d if r1 > r2 else c1 - r1 * (c2 - c1) / d
            return [punto_tang]
        else: return [] # d=0 solo posible si r1=r2, ya cubierto arriba

    # Dos puntos de intersección
    try:
        a = (r1**2 - r2**2 + d**2) / (2 * d)
        h_cuadrado = r1**2 - a**2
        if h_cuadrado < 0 and h_cuadrado > -epsilon : h_cuadrado = 0
        elif h_cuadrado < 0: return []
        h = np.sqrt(h_cuadrado)
        p2 = c1 + a * (c2 - c1) / d
        nx = (c2[1] - c1[1]) / d; ny = -(c2[0] - c1[0]) / d
        x3 = p2[0] + h * nx; y3 = p2[1] + h * ny
        x4 = p2[0] - h * nx; y4 = p2[1] - h * ny
        return [np.array([x3, y3]), np.array([x4, y4])]
    except (ZeroDivisionError, ValueError): return []

# ======================================================================

def rho_cech(MatrizDiscos, lambda_val, prec=1e-9):
    num_discos = len(MatrizDiscos)
    if num_discos <= 1: return -np.inf
    if num_discos == 2:
        puntos_interseccion = interseccion_dos_discos(MatrizDiscos[0], MatrizDiscos[1], lambda_val)
        return prec if puntos_interseccion else -np.inf

    max_min_rho_global = -np.inf
    for i in range(num_discos):
        for j in range(i + 1, num_discos):
            disco1 = MatrizDiscos[i]; disco2 = MatrizDiscos[j]
            puntos_interseccion = interseccion_dos_discos(disco1, disco2, lambda_val)
            if not puntos_interseccion: continue
            for punto_d in puntos_interseccion:
                min_rho_para_punto_d = np.inf
                valido_para_todos_k = True
                for k in range(num_discos):
                    if k == i or k == j: continue
                    ck = MatrizDiscos[k, :2]; rk_lambda = lambda_val * MatrizDiscos[k, 2]
                    if rk_lambda < 0: rk_lambda = 0 # Radios no negativos
                    dist_d_ck = np.linalg.norm(punto_d - ck)
                    rho_k = rk_lambda - dist_d_ck
                    if rho_k < -prec:
                        valido_para_todos_k = False; min_rho_para_punto_d = -np.inf; break
                    if rho_k < min_rho_para_punto_d: min_rho_para_punto_d = rho_k
                if valido_para_todos_k and min_rho_para_punto_d > max_min_rho_global:
                    max_min_rho_global = min_rho_para_punto_d
    return max_min_rho_global

# ======================================================================

def Bisection(f, a, b, tol=1e-8, max_iter=1000):
    if a > b: a, b = b, a
    try:
        fa = f(a)
        if abs(fa) < tol: return a, 0
        fb = f(b)
        if abs(fb) < tol: return b, 0
        if np.sign(fa) * np.sign(fb) >= 0:
            # print(f"Bisección: Mismos signos en [{a:.4f}, {b:.4f}] (fa={fa:.4e}, fb={fb:.4e}).")
            return np.nan, 0
    except Exception as e:
        print(f"Bisección: Error evaluando f en los extremos: {e}")
        return np.nan, 0

    iter_contador = 0
    while iter_contador < max_iter:
        iter_contador += 1
        m = a + (b - a) / 2
        try: fm = f(m)
        except Exception as e: print(f"Bisección: Error evaluando f({m}): {e}"); return np.nan, iter_contador

        if (b - a) / 2 < tol or abs(fm) < tol: return m, iter_contador
        if np.sign(fm) == np.sign(fa): a = m; fa = fm
        else: b = m
    # print(f"Bisección no convergió tras {max_iter} iteraciones.");
    return m, iter_contador

def ReglaFalsi(f, a, b, tol=1e-8, max_iter=1000):
    if a > b: a, b = b, a
    try:
        fa = f(a)
        if abs(fa) < tol: return a, 0
        fb = f(b)
        if abs(fb) < tol: return b, 0
        if np.sign(fa) * np.sign(fb) >= 0:
            # print(f"Regla Falsi: Mismos signos en [{a:.4f}, {b:.4f}] (fa={fa:.4e}, fb={fb:.4e}).")
            return np.nan, 0
    except Exception as e: print(f"Regla Falsi: Error evaluando f en extremos: {e}"); return np.nan, 0

    iter_contador = 0
    m_prev = b if abs(fa) < abs(fb) else a
    m = m_prev # Inicializar m

    while iter_contador < max_iter:
        iter_contador += 1
        try:
             denominador = fb - fa
             if abs(denominador) < 1e-15: raise ZeroDivisionError
             m = (a*fb - b*fa) / denominador
             # Asegurar que m esté dentro de [a, b] (puede salir por errores numéricos)
             if m < min(a,b) or m > max(a,b): m = a + (b-a)/2 # Fallback
        except (ZeroDivisionError, OverflowError): m = a + (b - a) / 2 # Fallback a bisección

        try: fm = f(m)
        except Exception as e: print(f"Regla Falsi: Error evaluando f({m}): {e}"); return np.nan, iter_contador

        if abs(m - m_prev) < tol or abs(fm) < tol: return m, iter_contador

        if np.sign(fm) == np.sign(fa):
            a = m; fa = fm
        else:
            b = m; fb = fm
        m_prev = m
    # print(f"Regla Falsi no convergió tras {max_iter} iteraciones.");
    return m, iter_contador

def Secante(f, x0, x1, tol=1e-8, max_iter=100):
    try:
        fx0 = f(x0)
        if abs(fx0) < tol: return x0, 0
        fx1 = f(x1)
        if abs(fx1) < tol: return x1, 1
    except Exception as e: print(f"Secante: Error evaluando f en puntos iniciales: {e}"); return np.nan, 0

    iter_contador = 0
    while iter_contador < max_iter:
        iter_contador += 1
        denominador = fx1 - fx0
        if abs(denominador) < 1e-15:
            # print("Secante: Denominador cercano a cero.");
            # Si fx1 es pequeño, hemos encontrado la raíz
            return x1 if abs(fx1) < tol else np.nan, iter_contador

        try:
             # Fórmula Secante
             x_next = x1 - fx1 * (x1 - x0) / denominador
        except (OverflowError, ZeroDivisionError):
             # print("Secante: Problema numérico.");
             return np.nan, iter_contador # Fallo numérico

        try: fx_next = f(x_next)
        except Exception as e: print(f"Secante: Error evaluando f({x_next}): {e}"); return np.nan, iter_contador

        if abs(x_next - x1) < tol or abs(fx_next) < tol: return x_next, iter_contador

        x0, fx0 = x1, fx1
        x1, fx1 = x_next, fx_next

    # print(f"Secante no convergió tras {max_iter} iteraciones.");
    return x_next, iter_contador

# ======================================================================

def encontrar_punto_clave_cech(MatrizDiscos, lambda_cech, prec=1e-9, tol=1e-6):
    num_discos = len(MatrizDiscos)
    if num_discos <= 2: return None
    best_min_rho = -np.inf
    punto_clave_candidato = None
    # Usar una tolerancia ligeramente mayor para encontrar el punto
    # ya que lambda_cech es una aproximación
    eval_prec = max(prec, tol*10)

    for i in range(num_discos):
        for j in range(i + 1, num_discos):
            disco1 = MatrizDiscos[i]; disco2 = MatrizDiscos[j]
            # Usar lambda_cech + epsilon pequeño puede ayudar a asegurar que estamos 'dentro'
            puntos_interseccion = interseccion_dos_discos(disco1, disco2, lambda_cech + eval_prec)
            if not puntos_interseccion:
                 # Probar también justo en lambda_cech por si la tangencia es la clave
                 puntos_interseccion = interseccion_dos_discos(disco1, disco2, lambda_cech)
            if not puntos_interseccion: continue

            for punto_d in puntos_interseccion:
                min_rho_para_punto_d = np.inf
                valido_para_todos_k = True
                for k in range(num_discos):
                    if k == i or k == j: continue
                    ck = MatrizDiscos[k, :2]; rk_lambda = lambda_cech * MatrizDiscos[k, 2]
                    if rk_lambda < 0: rk_lambda = 0
                    dist_d_ck = np.linalg.norm(punto_d - ck)
                    rho_k = rk_lambda - dist_d_ck
                    # Ser un poco más permisivo al verificar la contención
                    if rho_k < -eval_prec:
                        valido_para_todos_k = False; min_rho_para_punto_d = -np.inf; break
                    if rho_k < min_rho_para_punto_d: min_rho_para_punto_d = rho_k

                # Guardar el punto válido cuyo min_rho esté más cerca de 0 (o sea el menos negativo aceptable)
                if valido_para_todos_k:
                    # Queremos el punto cuyo min_rho esté más cerca de cero (idealmente >= -eval_prec)
                    # y sea el máximo entre los candidatos válidos.
                    if min_rho_para_punto_d >= -eval_prec and min_rho_para_punto_d > best_min_rho :
                          best_min_rho = min_rho_para_punto_d
                          punto_clave_candidato = punto_d
                    # Si no encontramos ninguno >= -eval_prec, tomamos el menos negativo
                    elif best_min_rho == -np.inf and min_rho_para_punto_d > best_min_rho:
                           best_min_rho = min_rho_para_punto_d
                           punto_clave_candidato = punto_d


    return punto_clave_candidato

# ======================================================================

def Cech_con_rho_y_punto(MatrizDiscos, lambda_vr, tol=1e-6, max_iter=100):
    num_discos = len(MatrizDiscos)
    if num_discos <= 2: return lambda_vr, None

    # Wrapper para pasar MatrizDiscos a rho_cech
    def rho_wrapper(lambda_val):
        if lambda_val <= 0: return -1e10 # Valor negativo grande
        # Usar precisión interna fina
        val = rho_cech(MatrizDiscos, lambda_val, prec=tol/100)
        # Asegurar que no sea NaN o Inf antes de devolver
        if not np.isfinite(val): return -1e10
        return val

    # --- Encontrar Intervalo [a, b] ---
    a = lambda_vr # Empezar en lambda_VR (o ligeramente antes?)
    # Podría ser mejor empezar 'a' un poco antes por si VR ya da rho>0 debido a errores numéricos
    a = max(0, lambda_vr * (1 - tol*10)) # Empezar un poco antes, pero no negativo
    rho_a = rho_wrapper(a)

    # Si rho(a) ya es positivo (o cero), entonces Cech <= a. Devolvemos 'a' (o lambda_vr).
    # Usar una pequeña ventana alrededor de 0 para la comprobación.
    if rho_a >= -tol:
        print(f"Advertencia: rho_cech({a:.4f})={rho_a:.4e} >= 0. Lambda Cech ≈ Lambda VR.")
        punto_clave_final = encontrar_punto_clave_cech(MatrizDiscos, a, prec=tol/10, tol=tol)
        return a, punto_clave_final # Devolver 'a' que podría ser ligeramente menor que VR

    # Buscar 'b' tal que rho(b) > 0
    b = lambda_vr * 1.5 if lambda_vr > tol else 1.0 # Empezar un 50% más allá o en 1
    rho_b = rho_wrapper(b)
    intentos = 0; max_intentos_intervalo = 30 # Aumentar límite por si acaso

    while rho_b <= tol and intentos < max_intentos_intervalo:
        if b < tol : b += 1.0 # Incremento aditivo si b cerca de 0
        elif b > 1e6: # Poner un límite superior para evitar overflow
             print("Error: Búsqueda de intervalo superó límite superior.")
             return np.nan, None
        else: b *= 1.5       # Incremento multiplicativo
        rho_b = rho_wrapper(b)
        intentos += 1

    if rho_b <= tol:
        print(f"Error: No se encontró intervalo [a, b] con cambio de signo para rho_cech (último b={b:.4f}, rho(b)={rho_b:.4e}).")
        return np.nan, None

    print(f"\n--- Cálculo de Lambda Cech (Raíz de rho) ---")
    print(f"Intervalo inicial para Bisección/F.Posición: [{a:.4f}, {b:.4f}]")

    # --- Aplicar los 3 Métodos ---
    results = {}
    lambda_bisec, iter_bisec = Bisection(rho_wrapper, a, b, tol, max_iter)
    results['Bisección'] = (lambda_bisec, iter_bisec)
    print(f"Bisección:       lambda ≈ {lambda_bisec if not np.isnan(lambda_bisec) else 'Fallo':.6f} (en {iter_bisec} iteraciones)")

    lambda_falsi, iter_falsi = ReglaFalsi(rho_wrapper, a, b, tol, max_iter)
    results['F. Posición'] = (lambda_falsi, iter_falsi)
    print(f"Falsa Posición:  lambda ≈ {lambda_falsi if not np.isnan(lambda_falsi) else 'Fallo':.6f} (en {iter_falsi} iteraciones)")

    # Puntos iniciales para Secante
    x0_sec = a
    x1_sec = a + (b - a) * 0.1 # Un punto cercano a 'a' dentro del intervalo
    if abs(x1_sec - x0_sec) < tol * 10: x1_sec = b # Usar b si están muy juntos
    print(f"Puntos iniciales para Secante: x0={x0_sec:.4f}, x1={x1_sec:.4f}")
    lambda_sec, iter_sec = Secante(rho_wrapper, x0_sec, x1_sec, tol, max_iter)
    results['Secante'] = (lambda_sec, iter_sec)
    print(f"Secante:         lambda ≈ {lambda_sec if not np.isnan(lambda_sec) else 'Fallo':.6f} (en {iter_sec} iteraciones)")

    # Seleccionar resultado final (priorizar Bisección si no falló)
    lambda_cech_final = np.nan
    if not np.isnan(lambda_bisec): lambda_cech_final = lambda_bisec
    elif not np.isnan(lambda_falsi): lambda_cech_final = lambda_falsi
    elif not np.isnan(lambda_sec): lambda_cech_final = lambda_sec

    if np.isnan(lambda_cech_final):
        print("Todos los métodos de búsqueda de raíces fallaron.")
        return np.nan, None

    # Asegurar Cech >= VR (aunque la búsqueda empezó en a >= VR*(1-eps))
    lambda_cech_final = max(lambda_vr, lambda_cech_final)

    # --- Encontrar el Punto Clave ---
    print(f"\nBuscando el punto clave asociado a lambda_Cech ≈ {lambda_cech_final:.6f}...")
    punto_clave_final = encontrar_punto_clave_cech(MatrizDiscos, lambda_cech_final, prec=tol/10, tol=tol)

    return lambda_cech_final, punto_clave_final

# ======================================================================
# === MAIN ===
# ======================================================================

# ==========
# Grafica un sistema de discos
# ==========

print("\n" + "="*50)
print("--- Parte 1: Grafica un sistema de discos ---")
print("="*50 + "\n")

# 1. Inicializamos el sistema
MatrizDiscos = IniciarSistemaDiscos(5)
print("\n--- Sistema de Discos Generado ---")
for i, d in enumerate(MatrizDiscos):
    print(f"Disco {i}: Centro=({d[0]:.2f}, {d[1]:.2f}), Radio={d[2]:.4f}")

# 2. Calculamos lambda_VR
lamda_vietorisRips = VietorisRips(MatrizDiscos)

# 3. Calculamos lambda_Cech y el punto clave con el nuevo método
lamda_cech, punto_clave = Cech_con_rho_y_punto(MatrizDiscos, lamda_vietorisRips)

# 4. Imprimimos los resultados
print("\n--- Resultado Final Complejos ---")
print(f"Lambda Vietoris-Rips: {lamda_vietorisRips:.6f}")

if not np.isnan(lamda_cech):
    print(f"Lambda Čech (raíz de rho): {lamda_cech:.6f}")
    if punto_clave is not None:
        print(f"Punto clave de intersección Čech: ({punto_clave[0]:.2f}, {punto_clave[1]:.2f})")
    else:
        print("Punto clave de intersección Čech: No encontrado (posible Cech=VR o fallo en encontrar_punto_clave_cech).")
else:
    print("Lambda Čech: No se pudo calcular.")

# --- Graficar ---
if not np.isnan(lamda_cech):
    MatrizDiscos_vr = MatrizDiscos.copy(); MatrizDiscos_vr[:,2] *= lamda_vietorisRips
    MatrizDiscos_cech_plot = MatrizDiscos.copy(); MatrizDiscos_cech_plot[:,2] *= lamda_cech

    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 7))

    DibujarDiscos(axes[0], MatrizDiscos, "1. Sistema Original")
    DibujarDiscos(axes[1], MatrizDiscos_vr, f"2. Sistema V-R (λ = {lamda_vietorisRips:.3f})")
    DibujarDiscos(axes[2], MatrizDiscos_cech_plot, f"3. Sistema Čech (λ = {lamda_cech:.3f}, raíz de rho)")

    # Graficar el punto clave si se encontró
    if punto_clave is not None:
        axes[2].plot(punto_clave[0], punto_clave[1], '*', color='red', markersize=15,
                     markeredgecolor='black', label='Punto Clave Čech')
        # Re-generar leyenda para asegurar que el punto clave esté
        handles, labels = axes[2].get_legend_handles_labels()
        # Filtrar handles/labels de discos existentes para evitar duplicados en la leyenda
        disk_labels = [f"D{i}:" for i in range(len(MatrizDiscos))]
        # Mantener solo handles cuya label comience con D0:, D1:, etc. O el punto clave
        final_handles = []
        point_handle_added = False
        for h, l in zip(handles, labels):
             is_disk = any(l.strip().startswith(dl) for dl in disk_labels)
             is_point = 'Punto Clave' in l
             if is_disk:
                 final_handles.append(h)
             elif is_point:
                 # Si el plot ya añadió el handle/label del punto, usarlo
                 final_handles.append(h)
                 point_handle_added = True

        # Si el plot no añadió el handle del punto a la leyenda (a veces pasa), crearlo manualmente
        if not point_handle_added:
             point_handle = plt.Line2D([0], [0], marker='*', color='red', markersize=15, markeredgecolor='black', linestyle='None', label='Punto Clave Čech')
             final_handles.append(point_handle)

        # Usar los handles filtrados y re-añadidos
        axes[2].legend(handles=final_handles, bbox_to_anchor=(1.05, 1), loc="upper left")


    axes[2].grid(True)
    plt.tight_layout(rect=[0, 0, 0.85, 1]) # Ajustar para leyenda
    plt.show()
else:
    print("\nNo se pudo calcular Lambda Čech, no se generará la gráfica.")


# ==========
# Analisis experimental
# ==========

print("\n" + "="*50)
print("--- Parte 2: Análisis Experimental Comparativo ---")
print("="*50 + "\n")

# --- Parámetros del Experimento ---
system_sizes = range(10, 101, 10) # N = 10, 20, ..., 100 (Originalmente era hasta N = 500)
num_instances_per_size = 25 # Número de sistemas aleatorios a generar por cada N (Originalmente eran 1000 pero se modifico a tan solo 25 por limitaciones computacionales)
tolerance = 1e-12 # Tolerancia para los métodos de búsqueda de raíces
max_iterations = 100

# Listas para almacenar los tiempos promedio
avg_times_bisec = []
avg_times_falsi = []
avg_times_secan = []
non_trivial_counts = [] # Contar cuántos casos no triviales encontramos por N

# --- Bucle Principal del Experimento ---
for N in system_sizes:
    print(f"Procesando sistemas de tamaño N = {N}...")
    times_bisec = []
    times_falsi = []
    times_secan = []
    non_trivial_count = 0

    for instance in range(num_instances_per_size):

        cx_temp = np.random.randint(-20, 21, size=N)
        cy_temp = np.random.randint(-20, 21, size=N)
        r_temp = np.random.uniform(1, 1.5, size=N)
        current_MatrizDiscos = np.column_stack((cx_temp, cy_temp, r_temp))

        # 2. Calcular Lambda VR
        lambda_vr_instance = VietorisRips(current_MatrizDiscos)

        # Wrapper para rho_cech con los discos actuales
        def current_rho_wrapper(lambda_val):
            if lambda_val <= 0: return -1e10
            # Usar precisión más fina internamente
            val = rho_cech(current_MatrizDiscos, lambda_val, prec=tolerance/100)
            return val if np.isfinite(val) else -1e10

        # 3. Filtrar casos triviales (rho(lambda_VR) >= 0)
        rho_at_vr = current_rho_wrapper(lambda_vr_instance)

        if rho_at_vr >= -tolerance: # Si es trivial, saltar a la siguiente instancia
            # print(f"  Instancia {instance+1}: Trivial (rho({lambda_vr_instance:.4f}) = {rho_at_vr:.4e} >= 0)")
            continue
        else: # Caso no trivial
            non_trivial_count += 1
            # print(f"  Instancia {instance+1}: No Trivial (rho({lambda_vr_instance:.4f}) = {rho_at_vr:.4e} < 0)")

            # 4. Encontrar intervalo [a, b] (usando el método de Cech_con_rho_y_punto)
            a = lambda_vr_instance
            b = a * np.sqrt(4.0/3.0)
            if abs(a-b) < tolerance: b = a + 1.0 # Asegurar intervalo si VR=0
            rho_b = current_rho_wrapper(b)
            intentos_b = 0; max_intentos_b = 20
            while rho_b <= tolerance and intentos_b < max_intentos_b:
                 if b < tolerance: b+= 1.0
                 elif b > 1e6: b=np.nan; break # Límite superior
                 else: b *= 1.5
                 rho_b = current_rho_wrapper(b)
                 intentos_b += 1

            if np.isnan(b) or rho_b <= tolerance:
                # print(f"    No se encontró intervalo para N={N}, instancia {instance+1}. Saltando.")
                non_trivial_count -= 1 # No contar si falló el intervalo
                continue # Saltar esta instancia si no se encuentra 'b'

            # 5. Medir tiempo para cada método en el intervalo [a, b]

            # Bisección
            start_time = time.time()
            lambda_b, it_b = Bisection(current_rho_wrapper, a, b, tolerance, max_iterations)
            end_time = time.time()
            if not np.isnan(lambda_b): times_bisec.append(end_time - start_time)

            # Falsa Posición
            start_time = time.time()
            lambda_f, it_f = ReglaFalsi(current_rho_wrapper, a, b, tolerance, max_iterations)
            end_time = time.time()
            if not np.isnan(lambda_f): times_falsi.append(end_time - start_time)

            # Secante
            x0_s = a
            x1_s = a + (b-a)*0.1 # Punto inicial cercano a 'a'
            if abs(x1_s - x0_s)< tolerance*10 : x1_s = b
            start_time = time.time()
            lambda_s, it_s = Secante(current_rho_wrapper, x0_s, x1_s, tolerance, max_iterations)
            end_time = time.time()
            if not np.isnan(lambda_s): times_secan.append(end_time - start_time)

    # 6. Calcular tiempos promedio para este N (evitando división por cero si no hubo casos no triviales)
    avg_times_bisec.append(np.mean(times_bisec) if times_bisec else np.nan)
    avg_times_falsi.append(np.mean(times_falsi) if times_falsi else np.nan)
    avg_times_secan.append(np.mean(times_secan) if times_secan else np.nan)
    non_trivial_counts.append(non_trivial_count)
    print(f"  Completado N={N}. Casos no triviales encontrados: {non_trivial_count}/{num_instances_per_size}")
    # print(f"    T_bisec={avg_times_bisec[-1]:.6f}s, T_falsi={avg_times_falsi[-1]:.6f}s, T_secan={avg_times_secan[-1]:.6f}s")


# --- Crear DataFrame con los resultados del experimento ---
results_df = pd.DataFrame({
    'N': list(system_sizes),
    'NonTrivialCases': non_trivial_counts,
    'AvgTime_Bisection': avg_times_bisec,
    'AvgTime_Falsi': avg_times_falsi,
    'AvgTime_Secante': avg_times_secan
})
results_df.set_index('N', inplace=True)

print("\n--- Tabla de Tiempos Promedio (segundos) ---")
pd.options.display.float_format = '{:.6e}'.format # Usar notación científica
print(results_df)

# --- Graficar los Tiempos Promedio ---
plt.figure(figsize=(12, 7))
plt.plot(results_df.index, results_df['AvgTime_Bisection'], 'o-', label='Bisección')
plt.plot(results_df.index, results_df['AvgTime_Falsi'], 's-', label='Regla Falsi')
plt.plot(results_df.index, results_df['AvgTime_Secante'], '^-', label='Secante')

plt.xlabel('Tamaño del Sistema (N)')
plt.ylabel('Tiempo Promedio de Ejecución (s)')
plt.title('Rendimiento Comparativo: Cálculo de Lambda Čech (Casos No Triviales)')
plt.legend()
plt.grid(True)
plt.yscale('log')
plt.show()

print("\n--- Fin del Análisis Experimental ---")
````

## Resultados y Conclusiones del Experimento

El objetivo de este análisis experimental fue comparar el rendimiento (tiempo de ejecución) de los métodos de Bisección, Regla Falsi y Secante para encontrar la escala de Čech ($\lambda_{\check{C}}$) en sistemas de discos no triviales, usando el enfoque de la función $\rho_M(\lambda) = 0$.

Los parámetros del experimento fueron:
* **Tamaños de Sistema (N):** De 10 a 100 discos, en incrementos de 10.
* **Instancias por Tamaño:** 25 instancias aleatorias.
* **Tolerancia (`tol`):** $10^{-8}$.
* **Filtrado:** Solo se midieron las instancias no triviales ($\rho_M(\lambda_{VR}) < 0$).
* **Intervalo de Búsqueda:** $[\nu_M, \sqrt{4/3}\nu_M]$.

### Observaciones de los Resultados

Los resultados de la tabla y la gráfica de rendimiento muestran dos puntos clave:

1.  **Crecimiento del Costo Computacional:** El tiempo de ejecución para todos los métodos aumenta significativamente con $N$. Esto es esperado, ya que la función `rho_cech` tiene una complejidad computacional alta (aproximadamente $O(N^3)$), volviéndose mucho más lenta de evaluar a medida que se añaden más discos.

2.  **Comparativa de Velocidad:**
    * **Bisección:** Fue consistentemente el método **más lento** en todos los tamaños de $N$.
    * **Falsa Posición y Secante:** Fueron notablemente más rápidos (en este experimento, hasta 10 veces más rápidos que Bisección). De los dos, el **Método de la Secante** demostró ser marginalmente el más rápido en promedio.

### Conclusión Final

Para este problema específico, existe un claro intercambio (trade-off) entre velocidad y robustez:

* **Bisección** es el método más **seguro y robusto**. Su convergencia está garantizada mientras exista un cambio de signo en el intervalo, pero es el más lento.
* **Secante y Falsa Posición** ofrecen una convergencia mucho más rápida. Sin embargo, son métodos numéricamente **menos estables**; son más sensibles a la forma de la función y a los puntos iniciales, y pueden tener problemas de convergencia (como saltar lejos de la raíz o fallar por divisiones cercanas a cero) en ciertos casos.

Si la **velocidad** es la prioridad y se pueden manejar fallos ocasionales, el **Método de la Secante** es la mejor elección. Si la **fiabilidad absoluta** es crítica para garantizar una respuesta en todos los casos no triviales, **Bisección** es la opción más fiable, aunque sea más costosa computacionalmente.
